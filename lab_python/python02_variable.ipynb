# Python Data Type(자료형)

* 숫자 타입
  *  정수: int(integer)
  *  실수: float(floating-point number)
* 문자 타입: str(string)
* 논리 타입: bool(boolean). 참(True)/거짓(False).

## 숫자 타입

```python
# 변수 선언, 값을 할당: 변수_이름 = 값
n = 123
print(n)  # 변수 n에 할당된 값을 출력.
print(type(n))  # 변수 n의 데이터 타입을 출력.
```

```python
pi = 3.141592
print(pi)
print(type(pi))
```

## 문자열 타입

```python
name = '오쌤'
print(name)
print(type(name))
```

## 논리 타입

```python
b = False  # 파이썬 참/거짓을 표현하는 예약어(식별자): True/False
print(b)
print(type(b))
```

```python
b = 1 > 5
print(b)
```

## 명시적 타입 변환

* `int(arg)`: 아규먼트(arg)로 전달된 값을 정수로 변환해서 반환.
  *  문자열 --> 정수
  *  실수 --> 정수
* `float(arg)`: 아규먼트로 전달된 값을 실수로 변환해서 반환.
  *  문자열 --> 실수
  *  정수 --> 실수
* `str(arg)`: 아규먼트로 전달된 값을 문자열로 변환해서 반환.
  *  정수, 실수, ... --> 문자열

```python
# input() 함수: 키보드에서 입력한 값들을 문자열로 받아들임.
x = int(input('정수 입력>>> '))  #> 입력값을 정수로 변환해서 변수 x에 저장.
y = int(input('정수 입력>>> '))

print(x - y)
```

# 연산자(Operator)

## 할당(assignment) 연산자

* `=`: 오른쪽의 값을 왼쪽의 변수에 할당(저장).

## 산술(arithmetic) 연산자

* 종류: `+, -, *, /, **, //, %`
* 산술 연산자는 일반적으로 숫자 타입(int, float)에서만 사용 가능.
* 일부 산술 연산자는 문자열과 같은 숫자 이외의 타입에서도 사용 가능.

```python
x, y = 10, 3
print('x + y =', x + y)
print('x - y =', x - y)
print('x * y =', x * y)
print('x / y =', x / y)
print('x ** y =', x ** y)  # 10의 3 거듭제곱
print('x // y =', x // y)  # 10을 3으로 나눈 몫
print('x % y =', x % y)  # 10을 3으로 나눈 나머지
```

문자열에서 사용할 수 있는 산술 연산자
* `문자열 + 문자열`: concatenate. 문자열 이어 붙이기.
* `문자열 * 정수`, `정수 * 문자열`: replicate. 문자열을 정수만큼 복사.

```python
print('Hello' + 'Python')
```

```python
print(3 * 'Hello')
print('Hello' * 3)
```

```python
x, y = '1', '2'
print(x + y)
```

## 복합 할당 연산자

* 종류: `+=, -=, *=, /=, ...`
* 산술 연산자와 할당 연산자가 합쳐져서 만들어진 연산자.

```python
salary = 100
print(salary)

# salary = salary * 1.1
salary *= 1.1
print(salary)
```

## 비교/논리 연산자

* 비교 연산자: `==, !=, >, >=, <, <=`
* 논리 연산자: `and, or, not`

```python
x, y = 100, 200
print(x == y)
print(x != y)
print(x > y)
print(x < y)
```

```python
print((x >= 50) and (x <= 150))
```

```python
x = None  # None: SQL에서 null과 같은 의미. 값이 없음.
print(x == None)
print(x != None)
#> 파이썬에서는 SQL처럼 is 연산자는 따로 없음.
```

# 연습문제

* 화씨 = (9/5) * 섭씨 + 32
* 키보드에서 섭씨 온도를 입력받아서, 화씨 온도로 변환된 값을 출력하세요.

```python
celsius = float(input('섭씨 온도 입력>>> '))
fahrenheit = (9 / 5) * celsius + 32
print('화씨 =', fahrenheit)
```
