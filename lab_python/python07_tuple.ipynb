# tuple

*   tuple(튜플): 저장된 아이템들을 변경할 수 없는 리스트.
*   인덱스 기반의 데이터 타입
    *   indexing: `tuple_name[index]`. 특정 인덱스 위치의 아이템을 읽어옴.
    *   slicing: `tuple_name[start:end]`. 특정 인덱스 범위의 아이템들로 이루어진 새로운 튜플. 잘라내기.
*   아이템을 추가, 삭제, 변경하는 기능(함수)들은 제공되지 않음.

```python
numbers = (1, 2, 3, 1, 2, 1)
print(type(numbers))
print(numbers)
```

```python
# tuple.count(value): 튜플에서 value가 등장하는 횟수.
numbers.count(1)
```

```python
# tuple.index(value): 튜플에서 value가 처음 나오는 인덱스.
numbers.index(1)
```
