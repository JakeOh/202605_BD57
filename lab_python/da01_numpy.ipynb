# Python `list` 클래스

*   여러 개의 값들을 저장할 수 있는 데이터 타입.
*   인덱스 기반 데이터 타입
    *   인덱스: 리스트에 아이템들이 저장된 순서(위치)를 표시하는 정수.
    *   indexing: 인덱스를 사용해서 리스트의 특정 위치의 값을 참조. (예) `list[0]`
    *   slicing: [start, end) 인덱스 범위의 아이템들을 잘라낸 리스트. (예) `list[0:5]`
*   산술 연산
    *   `list + list`: concatenate.
    *   `list * int, int * list`: replicate.

```python
list1 = [1, 2, 3]
list2 = [11, 22, 33]
```

```python
print(list1)  #> list.__str__() 메서드가 리턴하는 문자열
```

```python
list1  #> list.__repr__() 메서드가 리턴하는 문자열
```

```python
# indexing
print(list1[0])
print(list1[-1])
```

```python
# slicing
print(list1[:2])
print(list1[-2:])
```

```python
print(list1[:1])
```

```python
list1 + list2  # list1.extend(list2)와 동일.
```

```python
[x + y for x, y in zip(list1, list2)]
```

```python
list1 * 3
```

```python
[3 * x for x in list1]
```

# NumPy(Numerical Python)

*   `numpy.ndarray` 클래스: numpy 패키지의 가장 기본이 되는 클래스.
*   다차원 배열(n-dimensinal array)을 쉽고 빠르게 연산(+, -, *, /, ...)하기 위한 패키지.
*   numpy 패키지의 대부분의 기능들(연산자, 함수, 메서드 등)은 반복문의 기능을 가지고 있음.
    *   같은 인덱스의 원소들끼지(element-wise) 연산하는 기능을 가지고 있음.

```python
array1 = np.array([1, 2, 3, 4, 5])  # list를 ndarray로 변환해서 리턴
```

```python
type(array1)
```

```python
print(array1)  #> numpy.ndarray.__str__()
```

```python
array1  # numpy.ndarray.__repr__()
```

```python
array2 = np.array([10, 20, 30, 40, 50])
```

```python
# element-wise operation(같은 인덱스의 원소들끼리 연산)
print(array1 + array2)
print(array1 - array2)
print(array1 * array2)
print(array1 / array2)
```

```python
print(array1 + 1)
```

```python
print(array1 * 2)
```

# `numpy.ndarray` 클래스의 속성들(attributes)

```python
print(array1)
print('ndim =', array1.ndim)  # number of dimensions(차원 개수)
print('size =', array1.size)  # 크기(모든 원소들의 개수)
print('shape =', array1.shape)  # 모양(배열의 각 차원의 원소 개수들로 이루어진 튜플)
print('dtype =', array1.dtype)  # data type(배열 원소의 데이터 타입)
```

*   Python 숫자 타입
    *   정수: int
    *   실수: float

*   numpy 숫자 타입
    *   정수: int8, int16, int32, int64, ...
    *   실수: float16, float32, float64, ...

```python
array_2d = np.array([[1, 2, 3], [4, 5, 6]])
```

```python
print(array_2d)
```

```python
array_2d
```

```python
print('ndim =', array_2d.ndim)
print('size =', array_2d.size)
print('shape =', array_2d.shape)
print('dtype =', array_2d.dtype)
```

# `numpy.ndarray` 객체 생성 함수들

## `numpy.array(object)`

*   object를 numpy.ndarray 객체로 변환해서 리턴.
*   object: 배열과 비슷한 객체들. (예) list, tuple, ndarray, ...

```python
# 0 이상 10 미만의 정수들로 이루어진 ndarray
numbers = np.array([x for x in range(10)])
print(numbers)
```

## `numpy.arange(start, end, step)`

*   range를 사용해서 배열을 만들어 주는 함수
*   파이썬의 range() 함수 사용법과 같음.
*   1차원 배열을 생성.

```python
numbers = np.arange(10)  # 0 <= x < 10 범위의 1씩 증가하는 정수들의 배열
print(numbers)
```

```python
# 0 이상 10 미만의 짝수들을 갖는 배열
evens = np.arange(0, 10, 2)
print(evens)
```

## reshape

*   `numpy.reshape(array, newshape)` 함수(function)
*   `numpy.ndarray.reshape(newshape)` 메서드(method)

```python
array = np.arange(1, 13)
print(array)
```

```python
array2 = np.reshape(array, (3, 4))  # 함수 호출
print(array2)
```

```python
array3 = array.reshape((3, 4))  # 객체에서 메서드 호출
print(array3)
```

```python
array4 = array.reshape((4, 3))
print(array4)
```

```python
array4.reshape((12,))  # 2차원 배열 --> 1차원 배열
```

```python
# 메서드 연쇄 호출
array = np.arange(15).reshape((3, 5))
print(array)
```

## 난수 배열

### `numpy.random.rand(d0, d1, d2, ...)`

*   [0, 1) 범위의 균등분포(uniform distribution)를 따르는 난수들의 배열을 만듦.
*   아규먼트(d0, d1, ...)들은 각 차원의 원소 개수를 의미.

```python
array = np.random.rand(5)
print(array)
```

```python
array = np.random.rand(3, 5)
print(array)
```

### `numpy.random.randn(d0, d1, d2, ...)`

*   평균이 0이고 표준편차가 1인 표준정규분포(standard normal distribution)를 따르는 난수들의 배열을 만듦.
*   아규먼트(d0, d1, ...)들은 각 차원의 원소의 개수.

```python
array = np.random.randn(5)
print(array)
```

```python
array = np.random.randn(5, 3)
print(array)
```
