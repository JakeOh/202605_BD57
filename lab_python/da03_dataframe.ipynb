# 파생 변수

기존의 컬럼 값들을 사용해서 새로운 컬럼을 추가하는 것.

```python
tips.head()
```

```python
# 영수증 금액 대비 팁의 비율
tips.tip / tips.total_bill
```

```python
# 파생 변수 추가: data_frame['new_col_name'] = 계산식
tips['tip_ratio'] = tips.tip / tips.total_bill
```

```python
tips.head()
```

*   팁/영수증 비율(tip_ratio)의 평균
*   팁/영수증 비율이 평균보다 큰 행들 출력
*   팁/영수증 비율이 평균보다 큰 데이터에서 남성/여성의 비율

```python
tips.tip_ratio.mean()
```

```python
result = tips[tips.tip_ratio > tips.tip_ratio.mean()]
```

```python
result
```

```python
result.sex.value_counts()
```

```python
result.sex.value_counts(normalize=True)
```

```python
# 전체 남성/여성 고객들 중에서 팁-영수증 비율이 평균보다 많은 남성/여성 고객 비율
result.sex.value_counts() / tips.sex.value_counts()
```

titanic 데이터프레임의 컬럼들:

*   survived: 생존 여부(0=사망, 1=생존).
    *   alive 컬럼과 같은 내용('no'=사망, 'yes'=생존).
*   pclass: 객실 등급(1=First, 2=Second, 3=Third).
    *   class 컬럼과 같은 내용.
*   sex: 성별(male/female).
*   age: 나이.
*   who: 성인남성(man), 성인여성(woman), 어린이(child).
*   adult_male: 성인 남성 여부(True/False).
*   sibsp: 동승한 형제자매(siblings) 또는 배우자(spouses) 수.
*   parch: 동승한 부모(parents) 또는 자녀(children) 수.
*   fare: 운임(탑승권 가격).
*   embarked: 출발항(S, C, ...).
    *   embark_town(Southhampton, Cherbourg, ...).
*   deck: 객실 구역(A, B, C, ...).
