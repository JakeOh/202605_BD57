# 연습문제

*   숫자 타입 변수(컬럼)의 기술 통계량
*   카테고리 타입 변수의 빈도수
*   total_bill 컬럼의 값이 최댓값인 행을 출력
*   total_bill 컬럼의 값이 최솟값인 행을 출력
*   남성들의 tip의 평균
*   여성들의 tip의 평균
*   일요일의 tip의 평균
*   일요일 남성의 tip의 평균
*   일요일 여성의 tip의 평균

```python
tips.describe()  # 숫자 타입 변수들의 기술 통계량
```

```python
tips.sex.value_counts()  # 성별 빈도수
```

```python
tips['sex'].value_counts()
```

```python
tips.smoker.value_counts()  # 흡연여부 빈도수
```

```python
tips.day.value_counts()  # 요일 빈도수
```

```python
tips.time.value_counts()  # (식사) 시간 빈도수
```

```python
tips.size  #> DataFrame.size 속성: 행 개수 x 열 개수
```

```python
tips['size'].value_counts()  # 테이블 인원수 빈도수
```

```python
# 영수증 금액 최댓값
tips.total_bill.max()
```

```python
# 영수증 금액 최댓값인 행
tips[tips.total_bill == tips.total_bill.max()]
```

```python
# 영수증 금액 최솟값인 행
tips[tips.total_bill == tips.total_bill.min()]
```

```python
# 남성 팁의 평균
tips[tips.sex == 'Male'].tip.mean()
```

```python
tips.loc[tips.sex == 'Male', 'tip'].mean()
```

```python
# 여성 팁의 평균
tips[tips.sex == 'Female'].tip.mean()
```

```python
# 일요일 팁 평균
tips[tips.day == 'Sun'].tip.mean()
```

```python
# 일요일 남성 팁 평균
tips[(tips.day == 'Sun') & (tips.sex == 'Male')].tip.mean()
```

```python
# 일요일 여성 팁 평균
tips[(tips.day == 'Sun') & (tips.sex == 'Female')].tip.mean()
```
