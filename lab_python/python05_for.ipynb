# for 반복문

```
for 변수 in iterable_type:
    반복해서 실행할 코드(들)
```

*   iterable_type: 반복(iteration)이 가능한 데이터 타입.
    *   str, range, list, tuple, dict, ...
*   for 반복문은 iterable type에서 아이템들을 순서대로 하나씩 꺼내서 변수에 저장하고, 코드 블록을 실행.
*   for 반복문은 iterable type에서 더이상 꺼낼 아이템이 없을 때 종료됨.


````python
for x in '안녕하세요?':
    print(x)
````

# `for` 반복문과 `range()` 함수

````python
# range(start, end): start가 end보다 작은 경우, [start, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(1, 6):
    print(x)
````

````python
# range(end): [0, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(5):
    print(x)
````

````python
# range(start, end, step): start가 end보다 작고 step은 양의 정수인 경우,
# [start, end) 범위에서 step만큼씩 증가하는 정수들을 만듦.
for x in range(0, 10, 2):
    print(x)
````

````python
# range(start, end, step): start가 end보다 크고 step은 음의 정수인 경우,
# start >= x > end 범위에서 step만큼씩 감소하는 정수들을 만듦.
for x in range(10, 0, -1):
    print(x)
````

````python
for x in range(10, 0, -2):
    print(x)
````

````python
# for 반복문을 사용한 삼각형 그리기
for n in range(1, 6):
    print(n * '*')
````

````python
for n in range(5, 0, -1):
    print(n * '*')
````

# `break` vs `continue`

*   `break`: iteration 멈춤.
*   `continue`: 그 다음 iteration을 수행.

````python
for x in range(5):
    if x == 2:
        break
    print(x)
````

````python
for x in range(5):
    if x == 2:
        continue
    print(x)
````

for 반복문을 사용해서 구구단을 출력. 2단은 2x2까지, 3단 3x3까지, ..., 9단은 9x9까지 출력.

````python
for x in range(2, 10):
    print(f'=== {x}단 ===')
    for y in range(1, 10):
        print(f'{x} x {y} = {x * y}')
        if x == y:
            break
````

````python
for x in range(2, 10):
    print(f'=== {x}단 ===')
    for y in range(1, x + 1):
        print(f'{x} x {y} = {x * y}')
````

# `print()` 함수

*   escape character(탈출 문자열)
    *   `\n`: new line(줄바꿈)
    *   `\t`: tab(탭)
    *   `\'`: 작은따옴표. 문자열 안에서 인용부호를 포함하고 싶을 때.
    *   `\\`: 문자열 안에서 backslash(`\`)를 포함하고 싶을 때.

````python
print('안녕하세요\n파이썬')
````

````python
print('안녕하세요\t파이썬')
````

````python
print('I\'m a boy.')
````

````python
print('C:\\Users\\itwill')
````

`print()` 함수의 파라미터들:

*   `end`: print 함수가 아규먼트들을 출력한 후 마지막에 어떤 문자열을 추가할 것인 지를 설정.
    *   기본값은 줄바꿈(`\n`).
*   `sep`: separator(구분자). print 함수가 아규먼트들을 출력할 때, 각각의 아규먼트들을 구분하기 위한 문자열을 설정.
    *   기본값은 공백(' ') 1개.


````python
print('안녕하세요.')
print('점심은 맛있게 드셨나요?')
print('곧 퇴근 시간입니다.')
````

````python
print('안녕하세요.', end='::::')
print('점심은 맛있게 드셨나요?')
print('곧 퇴근 시간입니다.')
````

````python
print('안녕', '파이썬')
````

````python
print('안녕', '파이썬', sep='\t')
````
