# for 반복문

```
for 변수 in iterable_type:
    반복해서 실행할 코드(들)
```

*   iterable_type: 반복(iteration)이 가능한 데이터 타입.
    *   str, range, list, tuple, dict, ...
*   for 반복문은 iterable type에서 아이템들을 순서대로 하나씩 꺼내서 변수에 저장하고, 코드 블록을 실행.
*   for 반복문은 iterable type에서 더이상 꺼낼 아이템이 없을 때 종료됨.


````python
for x in '안녕하세요?':
    print(x)
````

# `for` 반복문과 `range()` 함수

````python
# range(start, end): start가 end보다 작은 경우, [start, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(1, 6):
    print(x)
````

````python
# range(end): [0, end) 범위의 1씩 증가하는 정수들을 만듦.
for x in range(5):
    print(x)
````
