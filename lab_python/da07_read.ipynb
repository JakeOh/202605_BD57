```python
import pandas as pd
```

# csv 파일의 첫번째 줄이 컬럼 이름인 경우

```python
# github에 저장된 csv 파일 경로
ex1_path = 'https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex1.csv'
```

```python
ex1_df = pd.read_csv(ex1_path)
```

```python
ex1_df
```

```python
ex1_df.info()
```

# cvs 파일의 첫번째 줄부터 데이터인 경우(컬럼 이름이 없는 경우)

```python
ex2_path = 'https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex2.csv'
```

```python
ex2_df = pd.read_csv(ex2_path,
                     names=['kor', 'eng', 'math', 'sci', 'msg'])
```

```python
ex2_df
```

```python
ex2_df.info()
```

`pandas.read_csv()` 함수의 names 파라미터는 DataFrame의 컬럼 레이블(이름)을 설정하기 위해서 사용.

*   csv 파일에 컬럼 레이블(이름)이 없는 경우. 파일의 첫번째 줄(행)부터 데이터인 경우.
*   csv 파일의 첫번째 행(줄)이 컬럼 이름인 경우에도, 컬럼 이름들을 원하는 이름으로 설정(재지정)하기 위해서 사용할 수도 있음.
    *   파일의 첫번째 줄(인덱스 0번 행)이 헤더임을 알려주기 위해서 `header=0` 파라미터와 함께 사용.

```python
ex1_df = pd.read_csv(ex1_path,
                     names=['kor', 'eng', 'math', 'sci', 'msg'],
                     header=0)
```

```python
ex1_df
```

```python
ex1_df.info()
```

# 값들이 공백으로 구분된 텍스트 파일

```python
ex3_path = 'https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex3.txt'
```

```python
ex3_df = pd.read_csv(ex3_path, sep='\\s+')  # sep와 delimiter는 같은 의미의 파라미터.
```

```python
ex3_df  #> ex3.csv 파일은 행 인덱스들을 가지고 있었던 파일 -> DataFrame에서도 행 인덱스들이 만들어짐.
```

```python
ex3_df.info()
```

# 컬럼 레이블과 데이터 이외의 주석 등을 포함하는 파일

```python
ex4_path = 'https://raw.githubusercontent.com/wesm/pydata-book/refs/heads/3rd-edition/examples/ex4.csv'
```

```python
ex4_df = pd.read_csv(ex4_path, skiprows=[0, 2, 3])
```

```python
ex4_df
```

```python
ex4_df.info()
```

*   csv(txt) 파일 --> DataFrame 생성: `pandas.read_csv()` 함수 사용.
*   DataFrame --> csv 파일 저장: `pandas.DataFrame.to_csv()` 메서드 사용.

```python
# ex4_df 데이터프레임에 파생변수(새로운 컬럼)을 추가.
ex4_df['temp'] = 'TEST'
ex4_df
```

```python
ex4_df.to_csv('ex4.csv')  #> 행 인덱스(레이블)도 csv 파일에 write됨.
```

```python
ex4_df.to_csv('ex4_noindex.csv', index=False)  # 행 인덱스는 파일에 write되지 않음.
```
