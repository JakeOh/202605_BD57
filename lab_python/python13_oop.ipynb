# OOP(Object-Oriented Programming, 객체지향 프로그래밍) 배경

````python
students = [
    {'id': 1, 'name': '홍길동', 'korean': 100, 'english': 50, 'math': 100},
    {'id': 2, 'name': '오쌤', 'korean': 50, 'english': 30, 'math': 60},
    {'id': 3, 'name': 'Scott', 'korean': 30, 'english': 100, 'math': 50},
]
````

````python
# 각 학생들의 번호, 이름, 총점, 평균을 출력.
for s in students:
    total = s['korean'] + s['english'] + s['math']
    mean = total / 3
    print(f'번호={s['id']}, 이름={s['name']}, 총점={total}, 평균={mean:.2f}')
````

위에서 작성한 코드의 문제점은, 각각의 학생을 의미하는 dict의 키가 모두 같을 때에만 정상적으로 동작하고, dict마다 키가 다르면(예: korean, korea) 에러가 발생하는 코드가 됨.

## 함수형 프로그랭(Functional Programming)

학생 정보를 저장하는 dict는 모두 동일한 키(key)를 갖도록 하기 위해서, 함수가 dict를 만들고 리턴하도록 해보면 어떨까?

````python
def create_student(id, name, korean, english, math):
    student = {
        'id': id,
        'name': name,
        'korean': korean,
        'english': english,
        'math': math,
    }

    return student
````

````python
stu1 = create_student(id=1, name='길동', korean=100, english=100, math=100)
stu1
````

````python
# create_student() 함수가 리턴하는 dict를 사용해서 총점을 계산해 주는 함수
def get_total(student):
    return student['korean'] + student['english'] + student['math']
````

````python
get_total(stu1)
````

````python
# create_student() 함수가 리턴하는 dict를 사용해서 평균을 계산해 주는 함수
def get_mean(student):
    return get_total(student) / 3
````

````python
get_mean(stu1)
````

create_student, get_total, get_mean 함수들을 사용한 성적 처리

````python
students = [
    create_student(1, '홍길동', 100, 80, 90),
    create_student(id=2, name='오쌤', korean=50, english=30, math=60),
    create_student(id=3, name='Scott', korean=50, english=100, math=95),
]
````

````python
students
````

````python
for s in students:
    print(f'번호={s['id']}, 이름={s['name']}, 총점={get_total(s)}, 평균={get_mean(s):.2f}')
````

get_total(student) 함수와 get_mean(student) 함수는 학생 dict 객체의 데이터와 매우 밀접한 관계를 가지고 있음.

데이터(국어, 영어, 수학 점수, ...)와 기능(총점 계산, 평균 계산, ...)이 매우 밀접하게 연결되어 있다.

# 객체지향 프로그래밍(OOP, Object-Oriented Programming)

*   **객체(object)**: 데이터(속성)과 기능(함수)을 가지고 있는 대상.
*   **클래스(class)**: 객체를 만들기 위한 설계도(코드).
    *   속성(property, attribute): 객체가 가져야하는 값(들), 변수(들).
    *   기능(function, **method**): 클래스 안에서 선언하는 함수.
        *   메서드를 선언할 때 항상 첫번째 파라미터는 `self`.
*   **인스턴스(instance)**: 메모리에 생성된 객체.
*   **생성자(constructor)**: 클래스 이름과 동일한 특별한 함수. 객체를 생성해 주는 함수.
    *   파이썬에서는 생성자를 호출하면 클래스에서 선언된 `__init__()` 메서드가 자동으로 호출됨.

````python
class Student:
    # 생성자 함수 - 객체 생성, 초기화.
    def __init__(self, id, name, korean, english, math):
        # 객체 속성들을 초기화(객체의 변수들에 값을 할당)
        self.id = id
        self.name = name
        self.korean = korean
        self.english = english
        self.math = math

    # 메서드 선언
    def get_total(self):
        """세 과목의 총점을 리턴."""
        return self.korean + self.english + self.math

    def get_mean(self):
        """세 과목의 평균을 리턴."""
        return self.get_total() / 3
````

변수, 함수, 클래스 이름 작성하는 권장 사항:
*   변수, 함수 이름 - snake 표기법. (예) student, student_id, get_total, ...
*   클래스 이름 - camel 표기법.  (예) Student, DataFrame, ...

````python
# Student 클래스의 생성자를 호출해서 Student 객체를 생성
student1 = Student(1, '홍길동', 100, 90, 95)
print(student1)  #> 학생 객체(Student object)

# Student 객체가 가지고 있는 속성들(properties), 변수들.
print(student1.id)
print(student1.name)
print(student1.korean)
print(student1.english)
print(student1.math)

# Student 객체가 가지고 있는 기능(메서드) 호출.
print('총점 =', student1.get_total())
print('평균 =', student1.get_mean())
````

````python
student2 = Student(id=2, name='오쌤', korean=60, english=30, math=50)
print(student2)
print('번호 =', student2.id)
print('이름 =', student2.name)
print('국어 =', student2.korean)
print('영어 =', student2.english)
print('수학 =', student2.math)
print('총점 =', student2.get_total())
print('평균 =', student2.get_mean())
````

````python
student3 = Student(id=3, name='Scott', korean=50, english=100, math=80)
````

````python
students = [student1, student2, student3]

for s in students:
    print(f'번호={s.id}, 이름={s.name}, 총점={s.get_total()}, 평균={s.get_mean():.2f}')
````

# 클래스 선언 & 메서드 호출 연습

## Ex. 1

원(Circle) 클래스
*   속성(property): 반지름(radius).
*   기능(method): 넓이(area) 계산, 둘레길이(perimeter) 계산.


````python
class Circle:
    # 생성자 - 변수 선언, 초기화
    def __init__(self, radius):
        self.radius = radius

    # 메서드
    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius
````

````python
c1 = Circle(1)  # 객체 생성
# print(type(c1))  #> class Circle

print('반지름 =', c1.radius)  # 생성된 객체의 속성 값을 읽어옴.
print('넓이 =', c1.area())  # 생성된 객체에서 메서드를 호출.
print('둘레길이 =', c1.perimeter())
````

````python
c2 = Circle(10)  # 반지름이 10인 원을 만듦.
print('반지름 =', c2.radius)  # 만들어진 원의 반지름을 출력.
print('넓이 =', c2.area())  # 만들어진 원의 넓이 출력.
print('둘레길이 =', c2.perimeter())  # 만들어진 원의 둘레 길이 출력.
````

## Ex. 2

직사각형(Rectangle) 클래스
*   속성: 가로(width), 세로(height) 길이.
*   기능: 넓이(area) 계산, 둘레길이(perimeter) 계산.


````python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)
````

````python
rect1 = Rectangle(3, 4)
print(type(rect1))
print(rect1)
print('가로 =', rect1.width)
print('세로 =', rect1.height)
print('넓이 =', rect1.area())
print('둘레 =', rect1.perimeter())
````

````python
rect2 = Rectangle(width=8, height=6)
print(type(rect2))
print(rect1)
print('가로 =', rect2.width)
print('세로 =', rect2.height)
print('넓이 =', rect2.area())
print('둘레 =', rect2.perimeter())
````

## Ex. 3

`Point` 클래스: 2차원 평면에서 한 점의 좌표를 표현
*   속성
    *   `x`: x 좌표 값. 기본값은 0.
    *   `y`: y 좌표 값. 기본값은 0.
*   기능
    *   `move`: 이동. 점의 위치를 원래 위치에서 x축의 방향으로 dx만큼, y축의 방향으로 dy만큼 이동.
    *   `distance`: 거리. 두 점 간의 거리를 계산해서 리턴.
        ```((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5```


````python
class Point:
    """2차원 평면에서 점의 좌표 (x, y)를 표현하는 클래스."""

    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y

    def move(self, dx=0, dy=0):
        """점의 좌표 (x, y)를 x축의 방향으로 dx만큼, y축의 방향으로 dy만큼 이동.
        이동된 점의 좌표는 (x+dx, y+dy).
        """
        self.x += dx
        self.y += dy

    # def distance(self, x, y):
    #     """다른 점의 x좌표와 y좌표를 전달받아서, 그 점까지의 거리를 계산.
    #     Args:
    #         x (float): 거리를 계산할 점의 x 좌표.
    #         y (float): 거리를 계산할 점의 y 좌표.
    #     """
    #     return ((self.x - x) ** 2 + (self.y - y) ** 2) ** 0.5

    def distance(self, other):
        """현재 점(self)에서 다른 점(other)까지의 거리를 리턴.

        Args:
            other (Point): (x, y) 좌표를 속성을 갖고 있는 Point 타입 객체.

        Returns:
            ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5
        """
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5
````

````python
p1 = Point()
````

````python
print(f'x={p1.x}, y={p1.y}')
````

````python
p1.move(dx=1, dy=2)
print(f'이동 후: x={p1.x}, y={p1.y}')
````

````python
p1.move(dy=-1)
print(f'이동 후: x={p1.x}, y={p1.y}')
````

````python
p2 = Point(5)  # positional argument
````

````python
print(f'x={p2.x}, y={p2.y}')
````

````python
p3 = Point(y=-1)  # keyword argument
````

````python
print(f'x={p3.x}, y={p3.y}')
````

````python
p4 = Point(x=1, y=1)  # keyword argument
````

````python
print(f'x={p4.x}, y={p4.y}')
````

````python
p3.distance(p4)
````

````python
p4.distance(p3)
````
