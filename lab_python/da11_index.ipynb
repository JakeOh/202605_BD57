## 계층적 인덱스(hierachical index, multi-level index)

```python
scores2 = pd.Series(data=np.random.randint(101, size=6),
                    name='score',
                    index=[[1, 1, 1, 2, 2, 2], [1, 2, 3, 1, 2, 3]])
scores2
```

```python
scores2.values
```

```python
scores2.index  #> MultiIndex 객체
```

Multi-level index에서 indexing과 slicing

*   첫번째 계층(level)의 인덱스만 가지고 indexing과 slicing을 할 수 있음.
*   두번째 계층(level)의 인덱스만 가지고는 indexing과 slicing은 할 수 없음.
*   튜플 형태의 인덱스로는 indexing과 slicing을 할 수 있음.


```python
scores2.index.nlevels
```

```python
scores2.loc[1]  # scores[1]
```

```python
scores2.loc[2]  # scores[2]
```

```python
scores2.loc[1:2]
```

```python
scores2[(1, 3)]  # indexing
```

```python
scores2[[(1, 3), (2, 3)]]  # fancy indexing
```

```python
scores2[(1, 2):(2, 2)]  # slicing
```

계층적 인덱스에서 인덱스의 순서 바꾸기

```python
scores2.swaplevel()
```

```python
# nlevels=3인 인덱스를 갖는 시리즈
np.random.seed(42)
s = pd.Series(data=np.random.rand(6),
              index=[np.arange(1, 7),
                     ['A'] * 3 + ['B'] * 3,
                     ['a', 'b', 'c'] * 2])
s
```

```python
s.loc[1:3]
```

```python
s.swaplevel()
# 파라미터 i=-2, j=-1 기본값. 끝에서 두번째 인덱스와 마지막 인덱스를 서로 바꿈(swap).
```

```python
s.swaplevel(i=1, j=2)
```

```python
s.swaplevel(i=0, j=1)  # 첫번째 인덱스와 두번째 인덱스를 서로 바꿈(swap).
```
