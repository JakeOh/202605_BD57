# Series

*   한 가지 타입의 값 여러 개를 저장하는 1차원 모양의 데이터 타입(클래스)
*   축(axis) 레이블을 가지고 있는 1차원 배열.
    *   정수 기반의 인덱스와 축 레이블을 함께 갖고 있음.

```python
s = pd.Series(data=[1, 2, -3, -4, 5])  # Series 클래스의 생성자 호출
```

```python
print(s)
```

```python
s
```

```python
s.index  # Series 객체의 인덱스
```

```python
s.values  # Series 객체의 값들(numpy.ndarray)
```

## indexing, slicing

*   `iloc`: numpy.ndarray에서 자동으로 부여되는 정수 인덱스를 기반으로 원소들을 참조.
    *   pandas.Series.values (ndarray 타입)가 가지고 있는 배열 인덱스를 사용.
    *   list 또는 ndarray에서 인덱스를 사용하는 방법과 같음.
    *   마지막 원소부터 거꾸로 부여되는 음수 인덱스를 사용할 수 있음.
*   `loc`: Series 객체를 생성할 때 설정하는 인덱스(레이블)를 기반으로 원소를 참조.
    *   ndarray(배열)의 인덱스와는 무관.
    *   마지막 원소부터 역순으로 자동 부여되는 음수 인덱스는 없음!

```python
s.iloc[0]  # 시리즈에서 첫번째 원소
```

```python
s.iloc[1]
```

```python
s.iloc[-1]  # 시리즈에서 마지막 원소
```

```python
s.iloc[1:3]  # slicing
```

```python
s.iloc[:3]  # 시리즈에서 첫 3개 원소들을 잘라냄.
```

```python
s.iloc[-3:]  # 시리즈에서 마지막 3개의 원소들을 잘라냄.
```

```python
s.loc[0]
```

```python
s.loc[1]
```

```python
s.loc[4]
```

```python
# s.loc[-1]  #> 에러 발생
```

```python
s.loc[1:3]  #> iloc를 사용한 slicing과는 차이가 있음!
```

```python
s.loc[:3]
```

```python
s.loc[3:]
```

```python
s.loc[-2:]
```

`series[index]`와 같이 인덱스를 사용할 수 있음.

*   경우에 따라서는 `series.iloc[index]`의 의미로 사용될 수도 있음.
    *   미래의 버전에서는 사라질 기능이기 때문에 사용하지 않는 것을 권장.

## fancy indexing

*   `series.iloc[인덱스들의 리스트]`
*   `series.loc[인덱스들의 리스트]`
    *   `series[인덱스들의 리스트]`

```python
s[['a', 'x']]
```

```python
s.loc[['a', 'x']]
```

```python
s.iloc[[0, 2]]
```

## boolean indexing

*   `series[boolean 값들의 리스트]`
*   `series.loc[boolean 값들의 리스트]`
*   `series.iloc[boolean 들]`는 사용 불가!

```python
s[[True, True, False, False, True]]
```

```python
s > 0
#> numpy.ndarray와 pandas.Series는 모두 element-wise 연산 기능을 제공.
```

```python
# 시리즈 s의 원소들 중에서 양수들만 선택.
s[s > 0]
```

```python
s[s < 0]
```

# DataFrame

*   테이블 모양의 2차원 데이터 타입.
*   행(row)과 열(column)을 가지고 있는 데이터 타입.

```python
data = {
    'city': ['서울'] * 3 + ['경기도'] * 3,
    'year': [2023, 2024, 2025] * 2,
    'pop': [0.91, 0.94, 0.93, 1.1, 1.2, 1.15],
}
```

```python
data
```

```python
df = pd.DataFrame(data=data)  # DataFrame 생성자 호출 --> DataFrame 객체 생성.
```

```python
print(df)
```

```python
df
```

## DataFrame 객체의 속성들(attributes)

```python
print(df.shape)  # shape: (row 개수, column 개수) 튜플
```

```python
print(df.index)  # index: 행 레이블
```

```python
print(df.columns)  # columns: 열 이름(레이블)
```

```python
# Index 객체(RangeIndex, Index, ...)들은 iterable --> for 반복문에서 사용 가능
for x in df.columns:
    print(x)
```

```python
print(df.values)  # values: 데이터프레임의 값들로만 만들어진 2차원 배열(ndarray)
```

## DataFrame에서 column을 선택하는 방법

*   `data_frame['column_name']`
    *   항상 사용 가능
*   `data_frame.column_name`
    *   컬럼 이름이 변수 이름 작성 문법에 맞지 않는 경우에는 사용할 수 없음.
        *   변수 이름은 영문자, 숫자, 밑줄(_)만 사용 가능.
        *   변수 이름은 숫자로 시작할 수 없음.
        *   파이썬 키워드(if, for, in, while, def, class, ...)들은 변수 이름으로 사용할 수 없음.
    *   컬럼 이름이 DataFrame 클래스의 속성(변수) 또는 메서드 이름과 같은 경우에는 사용할 수 없음.

```python
df['city']
```

```python
df.city
```

```python
df['pop']
```

```python
df.pop  #> DataFrame 클래스에서 선언된 메서드 이름
```

```python
df[['city', 'pop']]  # 2개 이상의 컬럼을 함께 선택
```

```python
df[['year', 'city', 'pop']]
```
