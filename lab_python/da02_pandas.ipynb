# Series

*   한 가지 타입의 값 여러 개를 저장하는 1차원 모양의 데이터 타입(클래스)
*   축(axis) 레이블을 가지고 있는 1차원 배열.
    *   정수 기반의 인덱스와 축 레이블을 함께 갖고 있음.

```python
s = pd.Series(data=[1, 2, -3, -4, 5])  # Series 클래스의 생성자 호출
```

```python
print(s)
```

```python
s
```

```python
s.index  # Series 객체의 인덱스
```

```python
s.values  # Series 객체의 값들(numpy.ndarray)
```

## indexing, slicing

*   `iloc`: numpy.ndarray에서 자동으로 부여되는 정수 인덱스를 기반으로 원소들을 참조.
    *   pandas.Series.values (ndarray 타입)가 가지고 있는 배열 인덱스를 사용.
    *   list 또는 ndarray에서 인덱스를 사용하는 방법과 같음.
    *   마지막 원소부터 거꾸로 부여되는 음수 인덱스를 사용할 수 있음.
*   `loc`: Series 객체를 생성할 때 설정하는 인덱스(레이블)를 기반으로 원소를 참조.
    *   ndarray(배열)의 인덱스와는 무관.
    *   마지막 원소부터 역순으로 자동 부여되는 음수 인덱스는 없음!

```python
s.iloc[0]  # 시리즈에서 첫번째 원소
```

```python
s.iloc[1]
```

```python
s.iloc[-1]  # 시리즈에서 마지막 원소
```

```python
s.iloc[1:3]  # slicing
```

```python
s.iloc[:3]  # 시리즈에서 첫 3개 원소들을 잘라냄.
```

```python
s.iloc[-3:]  # 시리즈에서 마지막 3개의 원소들을 잘라냄.
```

```python
s.loc[0]
```

```python
s.loc[1]
```
