# dict

*   사전(dictionary) 형식의 데이터 타입
*   키(key)와 값(value)의 쌍으로 저장하는 데이터 타입
    *   list, tuple: 인덱스를 기반으로 값들을 저장하는 데이터 타입
    *   dict: 키를 기반으로 값들을 저장하는 데이터 타입
    *   dict 타입에서 키의 역할은 list나 tuple에서 인덱스의 역할과 비슷.
        *   dict에 저장된 1개의 값을 참조할 때 사용.
        *   키는 중복된 값을 허용하지 않음.
        *   list, tuple의 인덱스는 정수만 가능, dict의 키는 정수, 문자열 등이 가능.

```python
# 학생의 번호와 이름을 저장하는 dict
students = {100:'홍길동', 101:'오쌤', 200:'Scott', 201:'홍길동'}
print(type(students))
print(students)
```

```python
len(students)  #> dict에서는 (key, value) 쌍이 1개의 아이템.
```

```python
# dict_name[key]: key에 매핑된 value를 반환.
students[201]
```

```python
# students[300]
#--> dict에 존재하지 않는 키를 사용하면 KeyError가 발생.
```

```python
numbers = [1, 3, 5, 7, 9]
print(numbers[4])  #> 9
# print(numbers[5])  #> IndexError가 발생
```

# `dict`의 함수(메서드)

```python
# dict_name.keys(): dict가 가지고 있는 키들의 리스트를 반환.
students.keys()
```

```python
# dict_name.values(): dict가 가지고 있는 값들의 리스트를 반환.
students.values()
```

```python
# dict_name.items(): (key, value) 튜플들의 리스트를 반환.
students.items()
```
