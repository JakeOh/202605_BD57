# dict

*   ΉΓυΉιΕ(dictionary) ΊαΧΉΜζΉζα ΈΞ░Ήζ┤ΊΕ░ ΊΔΑΉηΖ
*   ΊΓν(key)ΉβΑ Ά░Τ(value)Ήζα ΉΝΞΉε╝Έκε ΉιΑΉηξΊΧαΈΛΦ ΈΞ░Ήζ┤ΊΕ░ ΊΔΑΉηΖ
    *   list, tuple: Ήζ╕ΈΞ▒ΉΛνΈξ╝ Ά╕░Έ░αΉε╝Έκε Ά░ΤΈΥνΉζΕ ΉιΑΉηξΊΧαΈΛΦ ΈΞ░Ήζ┤ΊΕ░ ΊΔΑΉηΖ
    *   dict: ΊΓνΈξ╝ Ά╕░Έ░αΉε╝Έκε Ά░ΤΈΥνΉζΕ ΉιΑΉηξΊΧαΈΛΦ ΈΞ░Ήζ┤ΊΕ░ ΊΔΑΉηΖ
    *   dict ΊΔΑΉηΖΉΩΡΉΕε ΊΓνΉζα ΉΩφΊΧιΉζΑ listΈΓα tupleΉΩΡΉΕε Ήζ╕ΈΞ▒ΉΛνΉζα ΉΩφΊΧιΆ│╝ Έ╣ΕΉΛ╖.
        *   dictΉΩΡ ΉιΑΉηξΈΡε 1Ά░εΉζα Ά░ΤΉζΕ Ή░╕Ήκ░ΊΧι ΈΧΝ ΉΓυΉγσ.
        *   ΊΓνΈΛΦ ΉνΣΈ│╡ΈΡε Ά░ΤΉζΕ ΊΩΙΉγσΊΧαΉπΑ ΉΧΛΉζΝ.
        *   list, tupleΉζα Ήζ╕ΈΞ▒ΉΛνΈΛΦ ΉιΧΉΙαΈπΝ Ά░ΑΈΛξ, dictΉζα ΊΓνΈΛΦ ΉιΧΉΙα, Έυ╕ΉηΡΉΩ┤ ΈΥ▒Ήζ┤ Ά░ΑΈΛξ.

```python
# ΊΧβΉΔζΉζα Έ▓ΙΊα╕ΉβΑ Ήζ┤ΈοΕΉζΕ ΉιΑΉηξΊΧαΈΛΦ dict
students = {100:'ΊβΞΆ╕╕ΈΠβ', 101:'ΉανΉΝν', 200:'Scott', 201:'ΊβΞΆ╕╕ΈΠβ'}
print(type(students))
print(students)
```

```python
len(students)  #> dictΉΩΡΉΕεΈΛΦ (key, value) ΉΝΞΉζ┤ 1Ά░εΉζα ΉΧΕΉζ┤ΊΖε.
```

```python
# dict_name[key]: keyΉΩΡ ΈπνΊΧΣΈΡε valueΈξ╝ Έ░αΊβα.
students[201]
```

```python
# students[300]
#--> dictΉΩΡ Ήκ┤ΉηυΊΧαΉπΑ ΉΧΛΈΛΦ ΊΓνΈξ╝ ΉΓυΉγσΊΧαΈσ┤ KeyErrorΆ░Α Έ░εΉΔζ.
```

```python
numbers = [1, 3, 5, 7, 9]
print(numbers[4])  #> 9
# print(numbers[5])  #> IndexErrorΆ░Α Έ░εΉΔζ
```

# `dict`Ήζα ΊΧρΉΙα(ΈσΦΉΕεΈΥε)

```python
# dict_name.keys(): dictΆ░Α Ά░ΑΉπΑΆ│ι ΉηΙΈΛΦ ΊΓνΈΥνΉζα ΈουΉΛνΊΛ╕Έξ╝ Έ░αΊβα.
students.keys()
```

```python
# dict_name.values(): dictΆ░Α Ά░ΑΉπΑΆ│ι ΉηΙΈΛΦ Ά░ΤΈΥνΉζα ΈουΉΛνΊΛ╕Έξ╝ Έ░αΊβα.
students.values()
```

```python
# dict_name.items(): (key, value) ΊΛεΊΦΝΈΥνΉζα ΈουΉΛνΊΛ╕Έξ╝ Έ░αΊβα.
students.items()
```

```python
# dict_name.get(key, default=None):
# keyΆ░Α dictΉΩΡ Ήκ┤ΉηυΊΧαΈΛΦ Ά▓╜Ήγ░ΉΩΡΈΛΦ keyΉΩΡ ΊΧ┤ΈΜ╣ΊΧαΈΛΦ valueΈξ╝ Έ░αΊβα.
# keyΆ░Α dictΉΩΡ Ήκ┤ΉηυΊΧαΉπΑ ΉΧΛΈΛΦ Ά▓╜Ήγ░ΉΩΡΈΛΦ default(=None)Έξ╝ Έ░αΊβα. KeyErrorΆ░Α Έ░εΉΔζΊΧαΉπΑ ΉΧΛΉζΝ.
# dict_name[key]ΉβΑ Έ╣ΕΉΛ╖. KeyErrorΆ░Α Έ░εΉΔζΊΧι ΉΙα ΉηΙΉζΝ.
students.get(201)
```

```python
students.get(300)  #> keyΆ░Α dictΉΩΡ ΉΩΗΉΨ┤ΈΠΕ KeyErrorΆ░Α Έ░εΉΔζΊΧαΉπΑ ΉΧΛΉζΝ.
```

```python
students.get(300, 'No Name')
#> ΊΓν 300Ήζ┤ ΉηΙΉε╝Έσ┤ Ά╖╕ ΊΓνΉΩΡ ΊΧ┤ΈΜ╣ΊΧαΈΛΦ Ά░ΤΉζΕ Έ░αΊβα, ΉΩΗΉε╝Έσ┤ 'No Name'ΉζΕ Έ░αΊβα.
```

`dict_name[key] = value`

*   keyΆ░Α dictΉΩΡ ΉηΙΈΛΦ Ά▓╜Ήγ░, keyΉΩΡ ΈπνΊΧΣΈΡε Ά░ΤΉζΕ ΉΔΙΈκεΉγ┤ Ά░ΤΉε╝Έκε Έ│ΑΆ▓╜.
*   keyΆ░Α dictΉΩΡ ΉΩΗΈΛΦ Ά▓╜Ήγ░, (key, value) ΉΧΕΉζ┤ΊΖεΉζΕ dictΉΩΡ Ή╢ΦΆ░Α.

# dictΉβΑ for Έ░αΈ│╡Έυ╕

```python
# for-in Έ░αΈ│╡Έυ╕ΉΩΡΉΕε dictΈξ╝ ΉΓυΉγσΊΧαΈσ┤ keyΈΥνΉζΕ ΉΙεΉΕεΈΝΑΈκε Έ░αΈ│╡(iteration)ΊΧρ.
for k in students:
    print(k)
```

```python
for k in students:
    # print(k, students[k])
    print(k, students.get(k))
```

```python
for k, v in students.items():
    print(k, v)
```

```python
book = {'title': 'ΊΝΝΉζ┤ΉΞυ ΉαΙΉιε', 'author': 'ΉανΉΝν', 'price': 10_000}
print(book)
print(book['title'])
print(book.get('price'))
```

```python
for k, v in book.items():
    print(k, v)
```

# ΈΜνΉΨΣΊΧε ΊΔΑΉηΖΉζα ΉηΡΈμΝΈΥνΉζΕ ΉιΑΉηξΊΧαΈΛΦ dict

```python
contact = {
    'id': 1,
    'name': 'ΉανΉΝν',
    'phones': ['010-0000-0000', '02-0000-0000', '02-1234-5678'],
    'emails': {'company': 'jake@itwill.com', 'personal': 'jake@gmail.com'},
}
print(contact)
```

```python
contact['id']  # contact.get('id')
```

```python
contact['phones']
```

```python
# Ή▓τΈ▓ΙΉπ╕ ΉιΕΊβΦΈ▓ΙΊα╕
contact['phones'][0]
```

```python
contact.get('phones')[0]
```

```python
# company email
contact['emails']['company']
```

```python
contact.get('emails')['company']
```

```python
contact.get('emails').get('company')
```
