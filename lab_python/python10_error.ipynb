문법적인 오류가 아니라 예외적인 상황이 발생할 가능성이 있는 코드들은 예외 처리를 할 수 있음.

# try-except 구문

```
try:
    (1) 실행할 코드 블록
except [에러이름 [as 별명]]:
    (2) 예외 상황이 발생했을 때 예외를 처리하기 위한 코드 블록
[
else:
    (3) try 블록의 모든 코드들이 예외(오류) 없이 정상 수행이 됐을 때 실행할 코드 블록
finally:
    (4) 예외(오류) 발생 여부와 상관없이 항상 실행할 코드 블록
]
```

*   try 구문의 실행 순서
    *   예외(오류)가 발생하지 않은 경우: (1) --> (3) --> (4)
    *   예외(오류)가 발생한 경우: (1) --> (2) --> (4)
*   except 절에서 에러 이름을 생략하면, 모든 종류의 에러들을 처리할 수 있음.
*   except 절에서 에러 이름을 지정하면, 선언된 에러들만 처리할 수 있음.
*   처리해야 할 에러의 종류들이 여러개인 경우에는 except 절을 여러개 사용할 수도 있음.


````python
# 정수 2개를 입력받고, 사칙연산의 결과를 출력
try:
    x = int(input('정수 x 입력>>> '))
    y = int(input('정수 y 입력>>> '))
    print(f'{x} + {y} = {x + y}')
    print(f'{x} - {y} = {x - y}')
    print(f'{x} x {y} = {x * y}')
    print(f'{x} / {y} = {x / y}')
except ValueError as e:
    print('x와 y는 정수로 입력해야 합니다.')
except ZeroDivisionError as e:
    print('0으로 나눌 수 없습니다.')
else:
    print('사칙 연산을 모두 수행했습니다.')
finally:
    print('코드 셀 종료...')
````

````python
numbers = [random.randrange(100) for _ in range(5)]
numbers
````

````python
# 인덱스 입력받고, 리스트 numbers에서 해당 인덱스의 값을 출력.
try:
    index = int(input('인덱스 입력>>> '))
    print(f'numbers[{index}] = {numbers[index]}')
except:  # 모든 에러를 처리
    print('잘못된 인덱스를 입력했습니다.')
````
