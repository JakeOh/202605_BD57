# Conditional Statement(조건문)

## else가 없는 조건문

```
if 조건식:
    조건식을 만족할 때(조건식이 참일 떄) 실행할 문장(들).
```

*   `if 조건식` 뒤에는 반드시 콜론(:)을 써야 됨.
*   if 블록(조건식이 참일 때 실행할 문장들)은 반드시 들여쓰기(indentation)를 해야 됨.
*   같은 블록 안에서 들여쓰기 크기는 일정해야 함.


````python
age = int(input('나이 입력>>> '))
if age >= 0:  # age 값이 0보다 크거나 같으면
    print('안녕하세요')
    print(f'{age}살이군요...')  # formatted-string
````

## else가 있는 조건문

```
if 조건식:
    조건식이 참일 때 실행할 코드(들)
else:
    조건식이 거짓일 때 실행할 코드(들)
```

````python
age = int(input('나이 입력>>> '))
if age >= 0:
    print(f'당신은 {age}살이군요...')
else:
    print('나이는 음수가 될 수 없어요...')
````

(연습문제) 정수를 입력받아서, 그 정수가 짝수인 지, 홀수인 지를 출력하세요.

````python
number = int(input('정수 입력>>> '))
if number % 2 == 0:
    # number를 2로 나눈 나머지가 0과 같으면(number가 짝수이면)
    print(f'{number} = 짝수')
else:
    # 그렇지 않으면(number가 홀수이면)
    print(f'{number} = 홀수')
````

Falsy values: 파이썬에서 False로 취급하는 값들.

*   숫자(int, float): 0
*   문자열(str): 빈 문자열(empty string, '')
*   None: 변수에 값이 할당(저장)되어 있지 않는 경우를 표현하는 파이썬 예약어(keyword).
*   빈 리스트(empty list), ...


````python
number = int(input('정수 입력>>> '))
if number % 2:  # 만약에 number를 2로 나눈 나머지가 있으면
    print('홀수')
else:  # 나머지가 없으면
    print('짝수')
````

## chained conditional statement

```
if 조건식1:
    (조건식1)이 참일 때 실행할 코드(들)
elif 조건식2:
    (조건식2)가 참일 때 실행할 코드(들)
...
else:
    위의 모든 조건들이 거짓일 때 실행할 코드(들)
```

*   `elif` 블록은 1개 이상이 있을 수 있음.
*   `else` 블록은 생략이 가능.
*   모든 블록에서 들여쓰기 크기는 같은 크기로 작성해야 함.


숫자(float)를 입력 받아서, 그 숫자가 양수인 지, 0인 지, 음수인 지를 출력하세요.

````python
number = float(input('숫자 입력>>> '))
if number > 0:
    print('양수')
elif number == 0:
    print('영')
else:
    print('음수')
````

*   세 과목(국어, 영어, 수학)의 시험 점수를 입력받고 저장.
*   세 과목 점수의 총점을 계산해서 출력.
*   세 과목 점수의 평균을 계산해서 출력.
*   세 과목 점수 평균이 90점 이상이면 A, 80점 이상이면 B, 70점 이상이면 C, 60점 이상이면 D, 60점 미만이면 F라고 출력.

````python
korean = int(input('국어 입력>>> '))
english = int(input('영어 입력>>> '))
math = int(input('수학 입력>>> '))

total = korean + english + math
print(f'총점: {total}')

mean = total / 3
print(f'평균: {mean}')

if mean >= 90:
    print('A')
elif mean >= 80:
    print('B')
elif mean >= 70:
    print('C')
elif mean >= 60:
    print('D')
else:
    print('F')
````

## 중첩 조건문

*   if, elif, else 블록 안에 또다른 조건문이 포함된 경우.
*   **들여쓰기(indentation)** 를 특히 주의!!!
    *   들여쓰기 방식에 따라서 전혀 다른 로직이 될 수도 있음.
*   중첩 조건문은 논리 연산자(and, or, not)를 사용하는 조건문 또는 elif 블록으로 바꿀 수도 있음.

````python
number = float(input('숫자 입력>>> '))
if number > 0:
    print('양수')
else:
    if number == 0:
        print('영')
    else:
        print('음수')
````

````python
number = float(input('숫자 입력>>> '))
if number >= 0:
    if number > 0:
        print('양수')
    else:
        print('영')
else:
    print('음수')
````

# match 문

```
match 변수(또는 식):
    case 패턴1:
        변수(또는 식)이 패턴1과 같은 경우 실행할 코드(들)
    case 패턴2:
        변수(또는 식)이 패턴2와 같은 경우 실행할 코드(들)
    ...
    case _:
        변수(또는 식)이 위의 모든 패턴과 일치하지 않는 경우 실행할 코드(들)
```

*   `case 패턴`에서 패턴은 1개의 값 또는 `value1 | value2 | ...`와 같이 사용 가능.
*   match 구문은 if - elif - else 구문으로 바꿀 수 있음.


````python
menu = int(input('메뉴 입력>>> '))
match menu:
    case 1:
        print('입금')
    case 2:
        print('출금')
    case _:
        print('알 수 없는 메뉴')
````

````python
# 0(일요일), 1(월요일), 2(화요일), ..., 6(토요일)
day = int(input('요일 선택>>> '))
match day:
    case 1 | 2 | 3 | 4 | 5:
        print('학원 출근')
    case 0 | 6:
        print('집에서 복습')
    case _:
        print('알 수 없음')
````

입력받은 정수가 짝수인 지, 홀수인 지를 출력하는 문장을 match - case를 사용해서 작성해 보세요.

````python
number = int(input('정수 입력>>> '))
match number % 2:
    case 0:
        print('짝수')
    case 1:
        print('홀수')
````

# 난수(random number)를 만드는 방법

````python
import random  # 난수 생성 기능을 가지고 있는 random 모듈 가져오기(임포트)
````

````python
# random.random(): 0 이상 1 미만([0, 1) 범위)의 실수 난수를 만들어 줌.
random.random()
````

````python
# random.randrange(start, stop): [start, stop) 범위의 정수 난수를 만들어 줌.
random.randrange(0, 10)
````

````python
# random.randrange(stop): [0, stop) 범위의 정수 난수를 만들어 줌.
random.randrange(3)  # 0, 1, 2 중에 임의의 숫자 1개
````
