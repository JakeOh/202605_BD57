# 리스트(list)

*   `list`: 여러 개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.
*   아이템(item), 원소(element): 리스트에 저장된 각각의 값.
*   인덱스(index): 리스트에 아이템들이 저장된 위치를 표시하는 정수.
    *   인덱스는 **0** 부터 시작.
    *   마지막 인덱스는 (아이템 개수 - 1).
    *   리스트의 끝에서부터 역순으로 인덱스를 매길 때는 *음수 인덱스*를 사용.
*   리스트는 아이템들을 추가, 삭제, 변경할 수 있음.
    *   튜플(tuple)과 다른 점!

```python
# 숫자들을 저장하는 리스트
numbers = [1, 2, 10, 20, -1, -10, 100]
print(type(numbers))
```

```python
print(numbers)
```

```python
len(numbers)  #> 리스트 numbers의 아이템 개수.
```

```python
# 파이썬의 리스트는 여러가지 타입의 값들을 하나의 변수에 저장할 수 있음.
# 리스트 안에 다른 리스트가 포함될 수도 있음.
values = [1, 2, '안녕', '파이썬', True, False, 3.14, [1, 2, 3]]
print(type(values))
print(values)
```

# 인덱싱(Indexing)

인덱싱(indexing): 인덱스를 사용해서 리스트의 아이템을 찾고 사용하는 방법.

(문법) `list_name[index]`

```python
numbers = [1, -2, -3, 4, 5, 100]
numbers
```

```python
numbers[0]  #> 리스트 numbers의 첫번째 아이템(인덱스 0번 위치의 아이템)
```

```python
numbers[5]  #> 리스트 numbers의 마지막 아이템(인덱스 5번 위치의 아이템)
```

```python
numbers[-1]  #> 음수 인덱스 - 리스트의 마지막 아이템.
```

# 슬라이싱(slicing)

(문법) `list_name[start:end]`

*   리스트에서 start <= index < end 범위의 아이템들을 잘라내는 방법.
*   콜론(:)은 생략할 수 없음.
*   시작 인덱스 start를 생략하면 start 인덱스는 0. 리스트의 첫번째 아이템부터 잘라냄.
*   종료 인덱스 end를 생략하면 리스트의 마지막 아이템까지 잘라냄.
*   슬라이싱(slicing) 결과는 *새로운 리스트* !!


```python
numbers
```

```python
numbers[1:4]
```

```python
# 리스트 numbers의 첫 3개 아이템들을 잘라냄.
numbers[0:3]
```

```python
numbers[:3]
```

```python
# 리스트 numbers의 끝에서 3개 아이템들을 잘라냄.
numbers[3:6]
```

```python
numbers[3:]
```

```python
numbers[-3:]
```

```python
numbers[:]
```

```python
numbers[:1]  #> 아이템 1개를 갖는 "리스트"
```

```python
numbers[0]  #> 숫자
```
