# while 반복문

```
[초기화 문장]
while 조건식:
    조건식을 만족하는 동안에 반복할 코드(들)
    [조건식의 내용을 변경할 수 있는 코드]
```

*   조건식을 만족하는 동안에 블록 내부의 코드들을 반복해서 실행함.
*   조건식을 만족하지 못하면(조건식이 False)가 되면 반복을 멈춤.
*   `while 조건식` 다음에는 반드시 콜론(:)을 사용해야 함.
*   블록 안에서 들여쓰기의 크기는 일정하게 작성해야 함.


````python
# 1 ~ 5까지 순서대로 출력
n = 1  # 초기화 문장
while n <= 5:
    print(n)  # 반복할 문장
    n += 1  # n = n + 1
````

````python
# 5 ~ 1까지 내림차순으로 출력
n = 5
while n > 0:
    print(n)
    n -= 1
````

````python
# 2 4 6 8 10을 순서대로 출력
n = 2
while n <= 10:
    print(n)
    n += 2
````

````python
n = 1
while n <= 5:
    print(2 * n)
    n += 1
````

````python
n = 1
while n <= 10:
    if n % 2 == 0:
        print(n)
    n += 1
````
