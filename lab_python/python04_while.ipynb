# while 반복문

```
[초기화 문장]
while 조건식:
    조건식을 만족하는 동안에 반복할 코드(들)
    [조건식의 내용을 변경할 수 있는 코드]
```

*   조건식을 만족하는 동안에 블록 내부의 코드들을 반복해서 실행함.
*   조건식을 만족하지 못하면(조건식이 False)가 되면 반복을 멈춤.
*   `while 조건식` 다음에는 반드시 콜론(:)을 사용해야 함.
*   블록 안에서 들여쓰기의 크기는 일정하게 작성해야 함.


````python
# 1 ~ 5까지 순서대로 출력
n = 1  # 초기화 문장
while n <= 5:
    print(n)  # 반복할 문장
    n += 1  # n = n + 1
````

````python
# 5 ~ 1까지 내림차순으로 출력
n = 5
while n > 0:
    print(n)
    n -= 1
````

````python
# 2 4 6 8 10을 순서대로 출력
n = 2
while n <= 10:
    print(n)
    n += 2
````

````python
n = 1
while n <= 5:
    print(2 * n)
    n += 1
````

````python
n = 1
while n <= 10:
    if n % 2 == 0:
        print(n)
    n += 1
````

````python
# 9 7 5 3 1 순서로 출력
n = 9
while n > 0:
    print(n)
    n -= 2
````

````python
# 10보다 작은 자연수들 중에서 홀수들을 출력
n = 9
while n > 0:
    if n % 2:
        print(n)
    n -= 1
````

# `break`와 `continue`

*   `break`: 반복(iteration)을 종료. break가 포함된 가장 가까운 반복문만 빠져나감.
*   `continue`: 그 다음 반복(iteration)을 수행. continue 아래의 문장들은 건너뛰고, 그 다음 iteration 과정을 수행.

````python
n = 1
while n < 6:
    if n == 3:
        break
    print(n)
    n += 1
````

````python
n = 0
while n < 5:
    n += 1
    if n == 3:
        continue
    print(n)
````

## 문제 1. 구구단 2단 출력

```
2 x 1 = 2
2 x 2 = 4
...
2 x 9 = 18
```


````python
n = 1
while n < 10:
    print(f'2 x {n} = {2 * n}')
    n += 1
````

## 문제 2. 구구단 2단부터 9단까지 출력

```
=== 2단 ===
2 x 1 = 2
...
2 x 9 = 18
=== 3단 ===
3 x 1 = 3
...
3 x 9 = 27
...
```


````python
x = 2
while x < 10:
    print(f'=== {x}단 ===')
    # 구구단 내용 출력
    y = 1
    while y < 10:
        print(f'{x} x {y} = {x * y}')
        y += 1

    # 그 다음 구구단을 출력하기 위해서
    x += 1
````
