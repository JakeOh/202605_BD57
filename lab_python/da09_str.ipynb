자동차 이름(name)에서 첫번째 단어는 자동차 제조사 이름.

name 컬럼에서 첫번째 단어만 추출해서 company 파생 변수를 추가.

```python
# 문자열을 공백(' ')으로 잘라서 각각의 단어들로 이루어진 리스트의 첫번째 원소.
'chevrolet chevelle malibu'.split()[0]
```

```python
# 파생 변수 추가: data_frame['new_col_name'] = series/value
mpg['company'] = mpg.name.str.split().str[0]
```

```python
mpg
```

```python
companies = mpg.company.unique()  # 중복되지 않은 자동차 제조사 이름
companies
```

```python
companies.size  # 자동차 제조사 개수
```

```python
companies.sort()  # 오름차순 정렬
companies
```

## 데이터 정제(cleaning)

자동차 제조사 이름의 오타 수정:

*   chevroelt --> chevrolet
*   maxda --> mazda
*   mercedes --> mercedes-benz
*   toyouta --> toyota
*   vokswagen, vw --> volkswage


```python
mpg.loc[mpg.company == 'chevroelt', 'company'] = 'chevrolet'
mpg.loc[mpg.company == 'maxda', 'company'] = 'mazda'
mpg.loc[mpg.company == 'mercedes', 'company'] = 'mercedes-benz'
mpg.loc[mpg.company == 'toyouta', 'company'] = 'toyota'
mpg.loc[mpg.company.isin(['vokswagen', 'vw']), 'company'] = 'volkswagen'
```

```python
companies = mpg.company.unique()
companies.sort()
companies
```

```python
companies.size
```

## 자동차 제조사 빈도수

```python
company_counts = mpg.company.value_counts()
company_counts
```

```python
sns.countplot(data=mpg, y='company')
plt.show()
```

```python
plt.barh(y=company_counts.index, width=company_counts.values)
plt.show()
```

```python
company_counts2 = company_counts.sort_index(ascending=False)
company_counts2
```

```python
plt.barh(y=company_counts2.index, width=company_counts2.values)
plt.show()
```
