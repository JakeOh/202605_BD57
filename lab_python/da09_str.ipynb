```python
# 문자열 컬럼은 str 속성(StringMethods 타입)을 사용해서 문자열 메서드들을 호출할 수 있음.
gapminder.country.str.lower()
```

```python
gapminder.country.str.upper()
```

```python
# country 컬럼이 'Kor'을 포함하는 지 여부
gapminder.country.str.contains('Kor')
```

```python
# 'Kor'을 포함하는 행들
gapminder[gapminder.country.str.contains('Kor')]
```

```python
# 대/소문자 구분없이 'kor'을 포함하는 행들
gapminder[gapminder.country.str.lower().str.contains('kor')]
```

자동차 이름(name)에서 첫번째 단어는 자동차 제조사 이름.

name 컬럼에서 첫번째 단어만 추출해서 company 파생 변수를 추가.

```python
# 문자열을 공백(' ')으로 잘라서 각각의 단어들로 이루어진 리스트의 첫번째 원소.
'chevrolet chevelle malibu'.split()[0]
```

```python
# 파생 변수 추가: data_frame['new_col_name'] = series/value
mpg['company'] = mpg.name.str.split().str[0]
```
