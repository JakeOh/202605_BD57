# `str` 클래스 메서드

```python
msg = 'heLLo, pyTHon! hello PYTHON. 안녕하세요?'
```

```python
type(msg)
```

```python
print(msg)  #> str 클래스의 __str__ 메서드가 만들어 주는 문자열
```

```python
msg  #> str 클래스의 __repr__ 메서드가 만들어 주는 문자열
```

```python
result = msg.lower()  # 문자열 msg의 모든 영문자를 소문자로 변환한 "새로운" 문자열을 리턴.
print(result)
print(msg)  #> 원본 문자열은 변경되지 않음.
```

```python
result = msg.upper()
print(result)  #> 모든 영문자들을 대문자로 변환한 새로운 문자열을 리턴.
print(msg)  #> 원본 문자열은 변경되지 않음.
```

```python
msg.count('h')
```

```python
# 대/소문자 구분없이 'h' 또는 'H'의 개수
msg.lower().count('h')
```

```python
msg.find('l')  #> msg에서 'l'이 등장하는 가장 낮은 인덱스.
```

```python
msg.find('Q')  #> 찾으려는 문자열이 포함되어 있지 않은 경우에는 -1을 리턴.
```

```python
email = 'gildong@naver.com'

# 이메일 아이디(gildong)와 도메인(naver.com)을 각각 찾아서 출력.
idx = email.find('@')
print(idx)

print(email[:idx])  # 문자열에서 인덱스를 사용해서 slicing
print(email[(idx + 1):])
```

```python
result = msg.split()
#> 공백(space, tab, enter, ...)들을 구분자(separator)로 해서, 잘라진 문자열들의 리스트를 반환.
print(result)
```

```python
today = '2026-06-26'
result = today.split(sep='-')  #> '-'를 기준으로 문자열들을 자름.
print(result)
```

```python
print(email)
print(email.split(sep='@'))
print(email.split(sep='@')[0])
print(email.split(sep='@')[1])
```

```python
csv = '1,홍길동,100,90,80'  # comma-separated values
csv.split(sep=',')  #> 잘려진 문자열들의 리스트를 반환.
```
