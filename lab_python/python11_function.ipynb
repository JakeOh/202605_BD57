*   **함수(function)**: 기능을 수행하는 코드 블록.
*   **인수(argument)**: 함수를 호출할 때 함수에게 전달하는 값(들).
*   **매개변수(parameter)**: 아규먼트를 저장하기 위해서 함수를 정의(선언)하는 곳에서 선언하는 변수.
*   **반환값(return value)**: 함수가 기능을 수행하고 그 결과로 함수를 호출한 곳에 되돌려 주는 값.
    *   함수에 따라서 반환값이 있는 함수도 있고, 반환값이 없는 함수도 있음.

# 함수 선언(정의)

```
def function_name([parameter1, parameter, ...]):
    [""" 문서화 주석(documentation comment) """]
    함수가 수행할 코드(들)
    [return 반환값]
```

*   반환값이 없는 함수는 return 문장을 생략하거나, `return None` 문장을 사용할 수 있음.
*   `return`의 의미
    *   함수 종료
    *   함수를 호출한 곳에 값을 반환.


````python
# 함수 선언
def subtract(x, y):
    """숫자 x와 y를 전달받아서, x - y를 반환.

    Args:
        x (int, float): 뺄셈 피연산자.
        y (int, float): 뺄셈 피연산자.

    Returns:
        x - y
    """
    return x - y
````

````python
# 함수 호출
result = subtract(1, 2)
print(result)
````

````python
result = subtract(2, 1)
print(result)
````

## 값을 반환하지 않는 함수

````python
def repeat_message(msg, n):
    """
    메시지를 반복해서 출력.

    msg를 n번 반복해서 출력.

    Args:
        msg (str): 출력할 메시지.
        n (int): 출력할 횟수.
    """

    for _ in range(n):
        print(msg)

    # return None
````

````python
repeat_message('안녕하세요', 5)
````

````python
repeat_message('Hello, Python!', 3)
````

## 2개 이상의 값을 반환하는 함수

````python
def plus_and_minus(x, y):
    """
    두 숫자 x + y, x - y를 반환.

    Args:
        x (int, float): 덧셈과 뺄셈을 계산하기 위한 숫자.
        y (int, float): 덧셈과 뺄셈을 계산하기 위한 숫자.

    Returns:
        x + y, x - y
    """

    return x + y, x - y
````

````python
# 리턴값들을 저장하기 위해서 리턴값의 개수만큼 변수를 선언.
a, b = plus_and_minus(1, 2)
print(f'a = {a}, b = {b}')
````

````python
result = plus_and_minus(3, 2)
print(result)  #> 2개의 값(3+2, 3-2)을 갖는 튜플이 리턴.
print(result[0])  #> 튜플의 첫번째 아이템.
print(result[1])  #> 튜플의 인덱스 1번 아이템.
````
