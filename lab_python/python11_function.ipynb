# 함수 선언(정의)

```
def function_name([parameter1, parameter, ...]):
    [""" 문서화 주석(documentation comment) """]
    함수가 수행할 코드(들)
    [return 반환값]
```

*   반환값이 없는 함수는 return 문장을 생략하거나, `return None` 문장을 사용할 수 있음.
*   `return`의 의미
    *   함수 종료
    *   함수를 호출한 곳에 값을 반환.


````python
# 함수 선언
def subtract(x, y):
    """숫자 x와 y를 전달받아서, x - y를 반환.

    Args:
        x (int, float): 뺄셈 피연산자.
        y (int, float): 뺄셈 피연산자.

    Returns:
        x - y
    """
    return x - y
````

````python
# 함수 호출
result = subtract(1, 2)
print(result)
````

````python
result = subtract(2, 1)
print(result)
````

## 2개 이상의 값을 반환하는 함수

````python
def plus_and_minus(x, y):
    """
    두 숫자 x + y, x - y를 반환.

    Args:
        x (int, float): 덧셈과 뺄셈을 계산하기 위한 숫자.
        y (int, float): 덧셈과 뺄셈을 계산하기 위한 숫자.

    Returns:
        x + y, x - y
    """

    return x + y, x - y
````

## Ex. 1

*   함수 이름: make_list
*   기능: [start, end) 범위의 정수 난수 n개를 아이템으로 갖는 리스트를 반환.
*   파라미터: start(정수), end(정수), n(0 또는 양의 정수)
*   반환값: 정수 난수들의 리스트.


````python
def make_list(start, end, n):
    """
    [start, end) 범위의 정수 난수 n개를 갖는 리스트를 만들고 리턴.

    Args:
        start (int): 정수 난수를 생성하기 위한 시작 범위(포함).
        end (int): 정수 난수를 생성할 때 종료 범위(비포함). start보다 큰 정수.
        n (int): 반환할 리스트의 아이템 개수. 0 또는 양의 정수.

    Returns:
        정수 난수 n개를 갖는 리스트.
    """

    # numbers = []
    # for _ in range(n):
    #     numbers.append(random.randrange(start, end))
    numbers = [random.randrange(start, end) for _ in range(n)]

    return numbers
````

````python
rand_numbers = make_list(0, 10, 5)
rand_numbers
````

## Ex. 2

*   함수 이름: calculate_sum
*   기능: 숫자(정수 또는 실수)들을 원소로 갖는 리스트를 전달받아서, 그 리스트의 모든 원소들의 합을 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array의 모든 원소들의 합계.


````python
def calculate_sum(array):
    """
    숫자(정수 또는 실수)들을 아이템으로 갖는 리스트에서, 모든 원소들의 합을 계산하고 리턴.

    Args:
        array (list): 숫자(int, float)들의 리스트.

    Returns:
        array의 모든 원소들의 합.
    """

    total = 0
    for x in array:
        total += x

    return total
````

````python
total = calculate_sum(rand_numbers)
total
````

## Ex. 3

*   함수 이름: calculate_mean
*   기능: 숫자들을 원소로 갖는 리스트의 원소들의 평균을 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array의 모든 원소들의 평균.


````python
def calculate_mean(array):
    """
    숫자(정수, 실수)들의 리스트에서 모든 원소들의 평균을 계산하고 리턴.

    Args:
        array (list): 평균을 계산하고 싶은 숫자(int, float)들의 리스트.
                      원소가 하나도 없으면 ZeroDivisionError가 발생함.

    Returns:
        array의 원소들의 평균.
    """

    return calculate_sum(array) / len(array)
````

````python
calculate_mean(rand_numbers)
````

## Ex. 4

*   함수 이름: find_min_and_max
*   기능: 숫자들을 원소로 갖는 리스트에서 최솟값과 최댓값을 찾아서 리턴.
*   파라미터: array(숫자들의 리스트).
*   반환값: 리스트 array에서 최솟값과 최댓값.


````python
def find_min_and_max(array):
    """
    숫자(정수 또는 실수)들의 리스트에서 최솟값과 최댓값을 찾아서 리턴.

    Args:
        array (list): 숫자(int, float)들의 리스트. 적어도 1개 이상의 원소를 갖는 리스트.

    Returns:
        최솟값, 최댓값.
    """

    min = array[0]
    max = array[0]
    for x in array:
        if x < min:
            min = x
        if x > max:
            max = x

    return min, max
````

````python
min_val, max_val = find_min_and_max(rand_numbers)
print(min_val, max_val)
````

# Positional vs Keyword

````python
def make_point(x=0, y=0):
    print(f'Point({x}, {y})')
````

````python
make_point()
````

````python
make_point(1, 2)  #> positional argument
````

````python
make_point(1)  #> positional argument
````

````python
make_point(y=1)  #> keyword argument
````

함수를 호출할 때 아규먼트를 전달하는 방법:

*   **positional argument**: 선언된 파라미터 (위치) 순서대로 아규먼트들을 전달하는 방식.
*   **keyword argument**: `parameter=value` 형식으로 아규먼트를 전달하는 방식.
    *   키워드 아규먼트 방식은 선언된 파라미터 순서를 지키지 않아도 됨.
    *   **(주의)** 함수를 호출할 때 positional argument와 keyword argument를 함께 사용하는 경우에는, positional argument들을 먼저 써야 함.
    *   positional argument가 keyword argument보다 뒤에 나오면 문법 오류(SyntaxError).


````python
make_point(x=1, y=-1)
````

````python
make_point(y=3, x=-2)  #> 키워드 방식은 파라미터 순서를 지키지 않아도 됨.
````

````python
make_point(-1, y=3)  #> positional, keyword argument 순서 --> 가능.
````

````python
make_point(x=-1, 3)  #> keyword, postional argument 순서 --> 문법 오류.
````

# 가변길이 인수(variable-length arguments)

````python
print()  #> 아규먼트 0개
print('Hello')  #> 아규먼트 1개
print('Hello', 'Python')  #> 아규먼트 2개
print('hello', '점심 시간', '맛있게 드세요...')  # 아규먼트 3개
````

가변길이 인수:

*   함수를 호출할 때 아규먼트를 몇 개를 전달해도 됨.
    *   아규먼트를 하나도 전달하지 않아도 됨.
*   함수를 선언할 때 파라미터 이름 앞에 `*`를 사용하면 가변길이 인수가 됨.
*   함수 내부에서는 `*`가 있는 파라미터(가변길이 인수)를 튜플로 취급.
    *   가변길이 인수는 인덱스를 갖음.
    *   가변길이 인수는 for 반복문에서 사용할 수 있음.
*   `*`가 있는 파라미터(가변길이 인수)는 오직 1개만 선언할 수 있음.
*   가변길이 인수는 keyword 방식으로 아규먼트를 전달할 수 없음.
*   가변길이 인수 뒤에 선언된 파라미터(들)은 오직 keyword 방식으로만 아규먼트(들)을 전달할 수 있음.

`*`가 있는 파라미터(가변길이 인수)가 다른 파라미터보다 먼저 선언된 경우, 뒤에 선언된 파라미터들은 keyword-only(키워드 전용) 방식으로만 아규먼트를 전달할 수 있음.

````python
def calculate(*values, op):
    """
    values들의 합 또는 곱을 리턴.

    Args:
        values (int, float): 합 또는 곱을 계산하기 위한 숫자들.
        op (str): '+' 또는 '*'만 가능. 그 외의 값들은 ValueError 발생.

    Returns:
        op='+'인 경우, values들의 합계를 리턴.
        op='*'인 경우, values들의 곱을 리턴.
        values의 원소가 1개도 없는 경우 None을 리턴.

    Raises:
        op가 '+' 또는 '*'가 아닌 경우 ValueError가 발생.

    Examples:
        >>> calculate(1, 2, 3, 4, op='+')  # 1 + 2 + 3 + 4를 리턴
        10

        >>> calculate(1, 2, 3, 4, op='*')  # 1 * 2 * 3 * 4를 리턴
        24
    """

    if op not in ('+', '*'):  # if op != '+' and op != '*':
        raise ValueError("아규먼트 op는 '+' 또는 '*'만 가능.")
        # raise Error가 실행되면 함수 실행이 종료.

    # if not values:
    if len(values) == 0:
        result = None
    elif op == '+':
        result = 0
        for x in values:
            result += x
    else: # elif op == '*':
        result = 1
        for x in values:
            result *= x

    return result
````

````python
calculate(1, 2, 3, 4, op='+')
````

````python
calculate(1, 2, 3, 4, op='*')
````

````python
calculate(op='+')
````

````python
calculate(1, 2, 3, op='-')
````

# 가변길이 키워드 아규먼트(variable-length keyword arguments)

*   함수 파라미터를 선언할 때 `**`를 사용.
*   함수 호출할 때는 keyword 방식(`param1=value1, param2=value2, ...`)으로 아규먼트들을 전달.
    *   가변 길이 인수 --> 전달하는 아규먼트 개수는 제한이 없음.
    *   keyword 인수 --> 반드시 키워드와 함께 아규먼트를 전달해야 됨.
        *   키워드는 어떤 이름을 사용해도 됨.
*   함수 내부에서 가변길이 키워드 아규먼트는 dict 타입으로 취급.
    *   키워드는 dict의 키(key).
    *   아규먼트는 dict의 값(value).
*   가변길이 키워드 아규먼트(`**`가 있는 파라미터)는 오직 1개만 선언할 수 있음.
*   가변길이 아규먼트와 가변길이 키워드 아규먼트를 모두 갖는 경우, 가변길이 아규먼트를 먼저 선언해야 됨.

````python
def test_kwargs(*args, **kwargs):
    print(f'args={args}')
    print(f'kwargs={kwargs}')
````

````python
test_kwargs(1, 2, 'a', 'b')
````

````python
test_kwargs(1, 2, name='홍길동', age=16)
````

````python
def test_kwargs2(**kwargs, *args):
    pass
````

````python
def make_student_info(id, name, **kwargs):
    student = {'id': id, 'name': name}

    for k, v in kwargs.items():
        student[k] = v

    return student
````

````python
make_student_info(1, '홍길동')
````

# 함수 파라미터 선언에서 `/`와 `*`

*   함수 파라미터 선언에서 `/`는 **위치 전용 파라미터(positional-only parameter)의 끝**을 의미.
    *   `/` **왼쪽**의 파라미터들은 **위치 전용(positional-only)** 방식으로만 아규먼트를 전달할 수 있음.
    *   `/` 오른쪽의 파라미터들은 positional, keyword 방식 모두 가능.
*   함수 파라미터 선언에서 `*`는 **키워드 전용 파라미터(keyword-only parameter)의 시작**을 의미.
    *   `*` **오른쪽**의 파라미터들은 **키워드 전용(keyword-only)** 방식으로만 아규먼트를 전달할 수 있음.

````python
def db_config(url, /, user, password, *, timeout=30, retry_count=5):
    print(f'url={url}, user={user}, password={password}, timeout={timeout}, retry_count={retry_count}')
````

````python
db_config('localhost:1521:xe', 'scott', 'tiger')
````
