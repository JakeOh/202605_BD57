# 파일 쓰기(file write)

1. 파일 열기(open)
2. 파일 쓰기(write)
3. 파일 닫기(close)

---

`open(file, mode, ...)` 함수:

*   file 파라미터: 폴더(디렉토리) 이름을 포함한 파일 경로(이름).
*   mode 파라미터:
    *   'r': read. 읽기 전용 모드. 기본값.
    *   'w': write. 쓰기 모드. 기존에 작성된 모든 내용을 지우고 처음부터 새로 쓰기 시작.
    *   'a': append. 추가 모드. 기존에 작성된 내용 뒤에 이어 쓰기.
    *   't': text. 텍스트(문자열) 읽기/쓰기. 기본값.
    *   'b': binary. 바이너리(이진) 형식 파일(이미지, 동영상, ...) 읽기/쓰기.
*   파일 객체를 리턴
    *   읽기/쓰기/파일 닫기 등의 기능은 파일 객체가 가지고 있는 메서드를 사용.

```python
# 1. 파일 열기
f = open(file='hello.txt', mode='w')  # mode='wt'과 동일.

# 2. 파일 쓰기
f.write('Hello, Python!\n')
f.write('오늘은 금요일입니다. 불금 보내시나요?\n')
f.write('2026-06-26\n')

# 3. 파일 닫기
f.close()
```

*   파일을 open할 때, mode 파라미터를 'w' 또는 'a'로 설정할 경우:
    *   파일이 없으면 새로운 파일을 만듦.
    *   파일이 있으면 기존 파일을 열어줌.
        *   mode='w'일 때, 기존 내용이 모두 삭제됨.
        *   mode='a'일 때는 기존 내용이 유지.
*   파일을 open할 때 mode='r'로 설정한 경우:
    *   파일이 없으면 FileNotFoundError가 발생.

```python
f = open(file='hello.txt', mode='w')
f.close()
#> 기존 내용이 모두 삭제됨.
```

`with open(file, mode, ...) as 변수:` 구문을 사용하면 close 메서드가 자동으로 호출.

```python
with open(file='hello.txt', mode='w') as f:
    f.write('안녕...')
    f.write('Hello...\n')
    f.write('오늘 불금이네요...\n')
    # f.close()를 명시적으로 호출할 필요가 없음.
    # with-as 구문이 close() 메서드를 자동으로 호출함.
```

```python
with open(file='hello.txt', mode='a') as f:
    # mode='a'는 mode='at'와 동일.
    f.write('Hello, Python!\n')
    f.write('2026-06-26\n')
```

# 파일 읽기

1. 파일 열기(open 함수)
2. 파일에서 내용을 읽기(read, readline, ...)
3. 파일 닫기(close 메서드)

`with open(...) as ...` 구문을 사용하면 close 메서드는 자동으로 호출됨.

```python
with open(file='hello.txt') as f:
    # mode='r' 기본값.
    line = f.readline()
    print(line.strip())
    line = f.readline().strip()
    print(line)
```

`for variable in iterable:` 구문에서 파일 스트림 객체를 사용할 수 있음.

for 반복문 안에서는 readline() 메서드를 호출할 필요가 없음. for 반복문이 한 줄씩 읽어줌.

```python
with open(file='hello.txt') as f:
    for line in f:
        print(line.strip())
```

# Google Drive에서 파일 쓰기/읽기

Colab 런타임(가상머신)과 Google Drive를 연결(드라이브 마운트).

```python
# 구글 드라이브에 저장할 파일 경로(이름)
file_path = '/content/drive/MyDrive/Colab Notebooks/lab_python/emails.txt'
```

```python
# 구글 드라이브에 텍스트 파일을 생성, 내용 작성.
with open(file=file_path, mode='w') as f:
    f.write('jake@itwill.com\n')
    f.write('gildong@naver.com\n')
    f.write('hgd@yahoo.com\n')
```

```python
# 구글 드라이브에 저장된 텍스트 파일의 내용 읽기.
with open(file=file_path, mode='r') as f:
    for line in f:
        print(line.strip())
```
