# Ex. 2

*   1 이상 4 이하 범위의 정수 난수 10개를 저장하는 리스트 gender_codes를 만들고 출력하세요.
*   gender_codes의 값이 1 또는 3이면 'Male', 2 또는 4이면 'Female'을 저장하는 리스트 genders를 만들고 출력하세요.


````python
import random
````

````python
# gender_codes = []
# for _ in range(10):
#     gender_codes.append(random.randrange(1, 5))

gender_codes = [random.randrange(1, 5) for _ in range(10)]

print(gender_codes)
````

````python
genders = []
for x in gender_codes:
    # if x == 1 or x == 3:
    if x in (1, 3):
        genders.append('Male')
    else:
        genders.append('Female')

print(genders)
````

````python
# if-else 표현식을 list comprehension에서 사용.
genders2 = ['Male' if x in (1, 3) else 'Female'
            for x in gender_codes]
print(genders2)
````

# Ex. 4
아래의 두 2차원 리스트 m1, m2에서 같은 인덱스의 원소들끼리 뺄셈(m1 - m2) 결과를 저장하는 리스트를 만들고 출력하세요.

```
m1 = [
    [1, 2, 3],
    [4, 5, 6]
]
m2 = [
    [1, 1, 1],
    [2, 2, 2]
]

(결과)
result = [
    [0, 1, 2],
    [2, 3, 4]
]
```


````python
m1 = [
    [1, 2, 3],
    [4, 5, 6]
]
m2 = [
    [1, 1, 1],
    [2, 2, 2]
]

result = []
for row1, row2 in zip(m1, m2):
    minus = []
    for x, y in zip(row1, row2):
        minus.append(x - y)
    result.append(minus)

print(result)
````

````python
result2 = [[x - y for x, y in zip(row1, row2)]
           for row1, row2 in zip(m1, m2)]
print(result2)
````

````python
# scores = []
# for _ in range(5):
#     scores.append(random.randrange(1, 11))

num_of_items = 5

scores = [random.randrange(1, 11) for _ in range(num_of_items)]
print(scores)
````

# Ex. 8
문제 5에서 만든 리스트 scores의 원소들의 분산을 계산하고 출력하세요.

````python
# variances = []
# for x in scores:
#     variances.append((x - mean) ** 2)
variances = [(x - mean) ** 2 for x in scores]
print(variances)

sum_of_variances = 0
for x in variances:
    sum_of_variances += x

var = sum_of_variances / num_of_items
print(var)
````

# Ex. 10
문제 5에서 만든 리스트 scores에서 최댓값을 찾고 출력하세요.

````python
max = scores[0]  # 리스트 scores의 첫번째 원소를 최댓값이라고 가정.
for x in scores:  # 리스트 scores에서 첫번째 원소부터 마지막 원소까지 순서대로 반복하면서
    if x > max:  # 리스트에 있는 점수가 max보다 크다면
        max = x  # max를 그 점수로 바꿈.

print(max)
````

# Ex. 11
문제 5에서 만든 리스트 scores에서 최솟값을 찾고 출력하세요.

````python
min = scores[0]  # 리스트의 첫번째 원소를 최솟값으로 생각
for x in scores:  # 리스트의 모든 원소들 하나씩 순서대로 반복
    if x < min:  # 리스트에서 찾은 원소가 최솟값으로 가정한 값보다 작다면
        min = x  # 찾은 값을 최솟값으로 함.

print(min)
````

*   리스트의 모든 원소의 합계

$$
sum = \sum_{i = 1}^{n} x_i = x_1 + x_2 + ... + x_n
$$

*   리스트의 모든 원소의 평균(mean)

$$
\bar{x} = \frac{1}{n} \sum_{i = 1}^{n} x_i
$$

*   리스트의 모든 원소의 분산(variance)

$$
var = \frac{1}{n} \sum_{i = 1}^{n} (x_i - \bar{x})^2
$$

*   리스트의 모든 원소의 표준편차(standard deviation)

$$
std = \sqrt{var}
    = \sqrt{\frac{1}{n} \sum_{i = 1}^{n} (x_i - \bar{x})^2}
$$
