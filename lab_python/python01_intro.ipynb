Colab 코드셀 실행 방법:
* 실행 버튼 클릭.
* 런타임 메뉴에서 실행과 관련된 항목을 선택/클릭.
* 단축키 사용.
  * Ctrl+Enter: 현재 포커스(초점)가 있는 셀을 실행.
  * Shift+Enter: 현재 포커스가 있는 셀을 실행하고, 그 다음 셀로 포커스를 이동. 만약 다음 셀이 없으면 새로운 코드 셀을 자동으로 만들고 그 셀로 포커스를 이동.
  * Ctrl+F10: 현재 포커스가 있는 셀부터 노트북의 마지막 코드셀까지 모든 셀들을 실행.
  * Ctrl+F8: 현재 포커스가 있는 셀 위의 모든 코드 셀들을 순서대로 실행.
  * Ctrl+F9: 노트북의 처음부터 끝까지 모든 코드 셀들을 순서대로 실행.

```python
# 파이썬 인라인 주석(inline comment)
# 파이썬 문장은 문장 끝에 세미콜론(;)을 사용하지 않음!
# 파이썬에서 문자열(string)은 작은따옴표('')와 큰따옴표("")를 모두 사용할 수 있음.
# 파이썬의 식별자(예약어, 함수이름, 변수이름)들은 대소문자를 구분.

print('test')
print("안녕하세요.")
print("I'm a boy.")
print('I\'m a boy.')
```
