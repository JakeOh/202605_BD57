# 데이터프레임 합치기

*   merge: 두 개 이상의 데이터프레임을 공통된 컬럼(들)을 기준으로 합치는 것. SQL의 join과 비슷.
*   concat: 두 개 이상의 데이터프레임을 축(axis)을 따라서 합치는 것.

# merge

```python
emp = pd.DataFrame(data={
    'empno': [101, 102, 201, 202],
    'ename': ['Scott', 'Allen', 'King', '오쌤'],
    'deptno': [10, 20, 10, 40],
})
```

```python
emp
```

```python
dept = pd.DataFrame(data={
    'deptno': [10, 20, 30],
    'dname': ['IT', 'HR', 'Sales'],
})
```

```python
dept
```

*   `pandas.merge(left, right, how, on, ...)` 함수
*   `pandas.DataFrame.merge(right, how, on, ...)` 메서드
*   파라미터 `how`: merge 방식(inner, left, right, outer). 기본값은 inner.
*   파라미터 `on`: merge(join) 기준이 되는 컬럼(들의 리스트). merge하는 컬럼 이름이 같은 경우에는 생략 가능.

## inner join

```python
pd.merge(left=emp, right=dept)
```

```python
emp.merge(right=dept)
```

## left outer join

```python
pd.merge(left=emp, right=dept, how='left')
```

```python
emp.merge(right=dept, how='left')
```

## right outer join

```python
pd.merge(left=emp, right=dept, how='right')
```

```python
emp.merge(right=dept, how='right')
```

## full outer join

```python
pd.merge(left=emp, right=dept, how='outer')
```

```python
emp.merge(right=dept, how='outer')
```

## merge 기준 컬럼 이름이 다른 경우

```python
dept2 = pd.DataFrame(data={
    'dept_id': [10, 20, 30],
    'dept_name': ['아이티', '인사', '영업']
})
```

```python
dept2
```

```python
# inner join
emp.merge(right=dept2, left_on='deptno', right_on='dept_id')
```

```python
# left outer join
emp.merge(right=dept2, how='left', left_on='deptno', right_on='dept_id')
```

```python
# right outer join
emp.merge(right=dept2, how='right', left_on='deptno', right_on='dept_id')
```

```python
# full outer join
emp.merge(right=dept2, how='outer', left_on='deptno', right_on='dept_id')
```

# concat

## Series에서의 concat

```python
s1 = pd.Series(data=['scott', 'tiger', 'king'])
s1
```

```python
s2 = pd.Series(data=['홍길동', '오쌤'])
s2
```

```python
pd.concat([s1, s2])  #> s1과 s2의 인덱스가 유지.
```

```python
pd.concat([s1, s2], ignore_index=True)  #> s1과 s2의 인덱스들은 무시, RangeIndex가 새로 설정.
```

```python
s3 = pd.Series(data=['Sale', 'HR', 'IT'])
s3
```

```python
pd.concat([s1, s3], axis=1)
# axis=0: 기본값. 행(row) 방향으로 이어붙이기.
# axis=1: 열(column) 방향으로 이어붙이기.
```

# merge, concat 연습 2

미국 신생아 이름 데이터셋

*   https://github.com/wesm/pydata-book/tree/3rd-edition/datasets/babynames 데이터 파일들을 사용.
*   1880년부터 2010년까지 미국 신생아 이름, 성별, 출생자수가 정리된 파일들.
*   파일을 하나씩 읽어서 데이터프레임을 생성.
*   각각의 데이터프레임에 year 컬럼을 추가.
*   131개의 데이터프레임을 하나로 합치세요.(merge? concat?)
*   연도별 신생아 수를 찾고, 시각화.
*   연도별 성별 신생아 수를 찾고, 시각화.


```python
# 2010년 신생아 데이터
file_path = 'https://github.com/wesm/pydata-book/raw/refs/heads/3rd-edition/datasets/babynames/yob2010.txt'
df = pd.read_csv(file_path, names=['name', 'gender', 'births'])
df['year'] = 2010
```

```python
df.head()
```

```python
df.tail()
```

```python
yobs = []  # 각 연도의 txt 파일을 읽고 생성한 데이터프레임들을 저장하기 위한 리스트.
for year in range(1880, 2011):
    # txt 파일 경로
    file_path = f'https://github.com/wesm/pydata-book/raw/refs/heads/3rd-edition/datasets/babynames/yob{year}.txt'
    # txt 파일을 읽고 데이터프레임 생성
    df = pd.read_csv(file_path, names=['name', 'gender', 'births'])
    # year 파생변수 추가
    df['year'] = year
    # 리스트에 데이터프레임을 추가
    yobs.append(df)
```

```python
len(yobs)
```

```python
# 리스트의 첫번째 아이템
yobs[0]
```

```python
# 리스트의 마지막 아이템
yobs[-1]
```

```python
yob_df = pd.concat(yobs, ignore_index=True)
```

```python
yob_df.head(n=10)
```

```python
yob_df.tail(n=10)
```

```python
yob_df.info()
```

```python
yob_df.year.describe()
```

## 연도별 신생아 수

```python
yob_by_year = yob_df.groupby(by='year').births.sum()
yob_by_year
```

```python
sns.lineplot(x=yob_by_year.index, y=yob_by_year.values)
plt.grid()
plt.show()
```

```python
sns.lineplot(data=yob_df, x='year', y='births', estimator='sum', errorbar=None)
plt.grid()
plt.show()
```

## 연도별 성별 신생아 수

```python
yob_by_year_gender = yob_df.groupby(by=['year', 'gender']).births.sum()
yob_by_year_gender
```

```python
yob_by_year_gender.index  # pandas.Series.index 속성 -> MultiIndex
```

```python
yob_by_year_gender.values  # pandas.Series.values 속성.
```

```python
# Series의 index를 DataFrame의 컬럼으로 변환
yob_by_year_gender = yob_by_year_gender.reset_index()
yob_by_year_gender
```

```python
sns.lineplot(data=yob_by_year_gender, x='year', y='births', hue='gender')
plt.grid()
plt.show()
```
