# 데이터프레임 합치기

*   merge: 두 개 이상의 데이터프레임을 공통된 컬럼(들)을 기준으로 합치는 것. SQL의 join과 비슷.
*   concat: 두 개 이상의 데이터프레임을 축(axis)을 따라서 합치는 것.

# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# merge

```python
emp = pd.DataFrame(data={
    'empno': [101, 102, 201, 202],
    'ename': ['Scott', 'Allen', 'King', '오쌤'],
    'deptno': [10, 20, 10, 40],
})
```

```python
emp
```

```python
dept = pd.DataFrame(data={
    'deptno': [10, 20, 30],
    'dname': ['IT', 'HR', 'Sales'],
})
```

```python
dept
```

*   `pandas.merge(left, right, how, on, ...)` 함수
*   `pandas.DataFrame.merge(right, how, on, ...)` 메서드
*   파라미터 `how`: merge 방식(inner, left, right, outer). 기본값은 inner.
*   파라미터 `on`: merge(join) 기준이 되는 컬럼(들의 리스트). merge하는 컬럼 이름이 같은 경우에는 생략 가능.

## inner join

```python
pd.merge(left=emp, right=dept)
```

```python
emp.merge(right=dept)
```

## left outer join

```python
pd.merge(left=emp, right=dept, how='left')
```

```python
emp.merge(right=dept, how='left')
```

## right outer join

```python
pd.merge(left=emp, right=dept, how='right')
```

```python
emp.merge(right=dept, how='right')
```

## full outer join

```python
pd.merge(left=emp, right=dept, how='outer')
```

```python
emp.merge(right=dept, how='outer')
```

## merge 기준 컬럼 이름이 다른 경우

```python
dept2 = pd.DataFrame(data={
    'dept_id': [10, 20, 30],
    'dept_name': ['아이티', '인사', '영업']
})
```

```python
dept2
```

```python
# inner join
emp.merge(right=dept2, left_on='deptno', right_on='dept_id')
```

```python
# left outer join
emp.merge(right=dept2, how='left', left_on='deptno', right_on='dept_id')
```

```python
# right outer join
emp.merge(right=dept2, how='right', left_on='deptno', right_on='dept_id')
```

```python
# full outer join
emp.merge(right=dept2, how='outer', left_on='deptno', right_on='dept_id')
```

## 인덱스(row label)로 merge

```python
np.random.seed(42)
```

```python
df1 = pd.DataFrame(data={'value1': np.random.randint(10, size=6),
                         'value2': np.random.randint(10, size=6),},
                   index=['a', 'b', 'c'] * 2)
```

```python
df1
```

```python
df2 = pd.DataFrame(data={'data1': np.random.rand(3),
                         'data2': np.random.rand(3),},
                   index=['a', 'b', 'c'])
```

```python
df2
```

```python
pd.merge(left=df1, right=df2, left_index=True, right_index=True)
```

### 한쪽 데이터프레임은 컬럼으로, 다른 데이터프레임은 인덱스를 사용해서 merge

```python
df3 = pd.DataFrame(data={
    'value1': np.random.randint(10, size=6),
    'value2': np.random.randint(100, size=6),
    'value3': ['a', 'b', 'c'] * 2
})
```

```python
df3
```

```python
df4 = pd.DataFrame(data={'data1': np.random.rand(3),
                         'data2': np.arange(10, 40, 10),},
                   index=['a', 'b', 'c'])
```

```python
df4
```

```python
pd.merge(left=df3, right=df4, left_on='value3', right_index=True)
```

# concat

## Series에서의 concat

```python
s1 = pd.Series(data=['scott', 'tiger', 'king'])
s1
```

```python
s2 = pd.Series(data=['홍길동', '오쌤'])
s2
```

```python
pd.concat([s1, s2])  #> s1과 s2의 인덱스가 유지.
```

```python
pd.concat([s1, s2], ignore_index=True)  #> s1과 s2의 인덱스들은 무시, RangeIndex가 새로 설정.
```

```python
s3 = pd.Series(data=['Sale', 'HR', 'IT'])
s3
```

```python
pd.concat([s1, s3], axis=1)
# axis=0: 기본값. 행(row) 방향으로 이어붙이기.
# axis=1: 열(column) 방향으로 이어붙이기.
```

## DataFrame에서의 concat

```python
np.random.seed(42)
```
