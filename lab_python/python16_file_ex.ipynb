        `[['1', '1', '50', '98', '50'], ['2', '1', '60', '97', '60'], ...]`

*   위에서 만들어진 2차원 리스트는 문자열들을 저장하고 있는 리스트이기 때문에, 리스트의 모든 값들을 정수(int)로 변환

    `[[1, 1, 50, 98, 50], [2, 1, 60, 97, 60], ...]`

*   수학 점수들로만 이루어진 1차원 리스트를 만듦.

    `[50, 60, ...]`

*   수학 점수의 최댓값, 최솟값, 평균 출력.
*   1반 학생들의 수학 점수 총점, 평균, 최댓값, 최솟값 출력

```python
file_path = '/content/drive/MyDrive/Colab Notebooks/lab_python/exam.csv'
```

```python
with open(file=file_path) as f:
    line = f.readline()  # 파일의 첫번째 줄을 읽음.
    print(line.strip())

    # exam = []  # 데이터를 저장하기 위한 리스트.
    # for line in f:
    #     exam.append(line.strip().split(sep=','))

    # list comprehension
    exam = [line.strip().split(sep=',') for line in f]

    print(exam)
```

```python
# data = []  # 모든 문자열들을 int로 변환한 값들을 저장하기 위한 리스트.
# for row in exam:
#     temp = []  # 학생 1명의 데이터(id, class, math, english, science)를 저장하기 위한 리스트.
#     for x in row:
#         temp.append(int(x))
#     data.append(temp)

# list comprehension
data = [[int(x) for x in row] for row in exam]

print(data)
```

```python
# math = []  # 수학 점수들만 저장하기 위한 리스트.
# for row in data:
#     math.append(row[2])

math = [row[2] for row in data]

print(math)
```

```python
print('수학 최댓값 =', max(math))
print('수학 최솟값 =', min(math))
print('수학 평균 =', sum(math) / len(math))
```

```python
math_class1 = []  # 1반 학생들의 수학점수만 저장하기 위한 리스트.
for row in data:
    if row[1] == 1:  # class 값이 1이면
        math_class1.append(row[2])  # 수학 점수를 리스트에 추가.

print(math_class1)
```

```python
print('1반 수학 총점 =', sum(math_class1))
print('1반 수학 평균 =', sum(math_class1) / len(math_class1))
print('1반 수학 최댓값 =', max(math_class1))
print('1반 수학 최솟값 =', min(math_class1))
```

```python
# 반별 수학 평균 출력
for cls in range(1, 6):
    # math_by_class = []
    # for row in data:
    #     if row[1] == cls:
    #         math_by_class.append(row[2])
    math_by_class = [row[2] for row in data if row[1] == cls]
    mean = sum(math_by_class) / len(math_by_class)
    print(f'{cls}반 수학 평균 = {mean}')
```

## class 작성 연습

```python
class Student:
    def __init__(self, id, cls, math, english, science):
        self.id = int(id)
        self.cls = int(cls)
        self.math = int(math)
        self.english = int(english)
        self.science = int(science)

    def __repr__(self):
        return f'Student(id={self.id}, class={self.cls}, math={self.math}, english={self.english}, science={self.science})'
```

```python
Student('1', '1', '50', '98', '50')
```

```python
# 파일에서 읽은 내용을 Student 타입 객체로 생성해서 저장하기 위한 리스트.
exam_student = []

with open(file=file_path) as f:
    line = f.readline()
    for line in f:
        row = line.strip().split(sep=',')
        student = Student(id=row[0], cls=row[1],
                          math=row[2], english=row[3], science=row[4])
        exam_student.append(student)

exam_student
```

```python
math = [s.math for s in exam_student]
print(math)
```

```python
english = [s.english for s in exam_student]
print(english)
```

```python
science = [s.science for s in exam_student]
print(science)
```

```python
# 반별 과학 평균 출력
for c in range(1, 6):
    science_by_class = [s.science for s in exam_student if s.cls == c]
    # print(science_by_class)
    mean = sum(science_by_class) / len(science_by_class)
    print(f'{c}반 과학 평균 = {mean}')
```

## 클래스 작성 연습 2

```python
class MyDataFrame:
    def __init__(self, data, columns):
        self.data = data  # 2차원 리스트.
        self.columns = columns  # 1차원 리스트.

    def __repr__(self):
        result = '\t'.join(self.columns) + '\n'
        result += '-' * 40 + '\n'
        for row in self.data:
            result += '\t'.join(str(x) for x in row) + '\n'

        return result
```
