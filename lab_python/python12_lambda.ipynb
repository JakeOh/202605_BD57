# 1급 객체(first-class object)로서의 함수

**함수(function)는 객체(object)다.**

*   객체(object): 숫자, 문자열, 논리값, list, tuple, dict, ...
*   객체는 변수에 할당할 수 있음.
*   객체는 함수 아규먼트로 전달할 수 있음.
*   객체는 함수가 반환(return)할 수 있음.
*   객체는 함수 내부에서 만들고 사용할 수 있음.


````python
x = 1  # 정수 1을 변수 x에 할당 --> 객체를 변수에 할당.
print(x)  # 변수 x의 값을 출력 --> 객체를 print 함수의 아규먼트로 전달.
````

````python
x = len('hello')  # len() 함수의 리턴값을 변수 x에 할당. --> 함수는 객체를 리턴할 수 있음.
print(x)
````

````python
def double(x):
    return x * 2
````

````python
print(double(4))
#> double() 함수를 호출하고 그 리턴값을 출력.
````

````python
print(double)
#> double이라는 이름의 객체를 출력.
````

## 함수의 아규먼트로 함수를 전달

````python
def calculate(x, y, fn):
    result = fn(x, y)
    return result
````

````python
def plus(x, y):
    return x + y
````

````python
calculate(x=1, y=2, fn=plus)
````

````python
def minus(x, y):
    return x - y
````

````python
calculate(x=1, y=2, fn=minus)
````

````python
def is_greater(x, y):
    return x > y
````

````python
calculate(x=1, y=1, fn=is_greater)
````

## 내부 함수, 함수를 리턴하는 함수

내부 함수(inner function), 지역 함수(local function):

*   함수 안(내부)에서 선언된 함수.
*   선언된 함수(외부 함수) 안에서만 호출할 수 있음.
*   선언된 함수 바깥에서는 내부 함수 이름으로는 호출할 수 없음.
*   외부 함수의 지역 변수들(파라미터 포함)을 사용할 수 있음.
*   함수가 내부 함수 객체를 리턴하면, 외부 함수 바깥에서도 호출할 수 있음.


````python
# 외부 함수 선언
def make_incrementor(n):
    # 내부 함수 선언
    def add_n(x):
        return x + n

    # 함수 객체를 리턴
    return add_n
````

````python
add_n(10)
#> NameError 발생. 내부(지역) 함수 이름으로는 외부에서 호출할 수 없음.
````

````python
plus_two = make_incrementor(n=2)
print(plus_two)
````

````python
plus_two(x=10)
````

````python
plus_ten = make_incrementor(n=10)
plus_ten(x=10)
````

````python
make_incrementor(5)(10)
````

# 람다 표현식(Lambda Expression)

(문법) `lambda param1, param2, ...: expression(식)`

*   이름이 없는 함수.
*   함수 이름 선언 없이, 파라미터 선언과 리턴값 또는 리턴 식으로 함수를 선언하는 것.
*   람다 표현식은 변수에 할당할 수 있고, **함수의 아규먼트로 전달**할 수 있음.


````python
# 함수 선언
# def plus_one(x):
#     return x + 1

# lambda 표현식
plus_one = lambda x: x + 1
print(plus_one)  #> 함수(function) 객체

plus_one(11)  #> 함수 호출
````

````python
calculate(1, 2, lambda x, y: x + y)
````

````python
calculate(1, 2, lambda x, y: x - y)
````

````python
calculate(2, 3, lambda x, y: x * 3)
````

````python
calculate(3, 2, lambda x, y: x / y)
````

````python
calculate(1, 2, lambda x, y: x > y)
````

## lambda를 아규먼트로 전달하는 예


*   filtering: 조건을 만족하는 아이템들만 찾기.
    ```
    filter(function, iterable)
    ```
    *   iterable 안에 있는 원소들을 하나씩 순서대로 function의 아규먼트로 전달하고, function이 True를 리턴하는 원소들만 선택.
    *   function은 아규먼트가 1개이고, True/False(bool)를 리턴하는 함수여야 함.
*   mapping: 아이템들을 일정한 규칙에 따라서 다른 값으로 변환(매핑).
    ```
    map(function, iterable)
    ```
    *   iterable 안에 있는 원소들을 하나씩 순서대로 function의 아규먼트로 전달하고, function이 리턴하는 값들로 매핑.
    *   function은 아규먼트가 1개이고, 리턴값이 있는 함수여야 함.
*   `filter`, `map` 함수는 list 또는 tuple과 같은 객체를 리턴하지는 않음.
    *   필터링 또는 매핑된 결과를 리스트로 만들려면, filter 또는 map의 리턴 값을 `list` 함수의 아규먼트로 전달하면 됨.


````python
import random
````

````python
numbers = [random.randrange(100) for _ in range(10)]
numbers
````

````python
# numbers의 아이템들 중에서 짝수들만 필터링.
[x for x in numbers if x % 2 == 0]
````

````python
result = filter(lambda x: x % 2 == 0, numbers)
list(result)  # 필터링된 결과를 list 객체로 만듦.
````

````python
# filter 함수를 사용해서 numbers에서 홀수들만 필터링.
result = filter(lambda x: x % 2 == 1, numbers)
list(result)
````

````python
# numbers의 원소가 짝수이면 'even', 홀수이면 'odd'로 매핑
['even' if x % 2 == 0 else 'odd' for x in numbers]
````

````python
result = map(lambda x: 'even' if x % 2 == 0 else 'odd', numbers)
list(result)  #> 매핑된 결과를 리스트로 만듦.
````

numbers의 원소가 0 ~ 19이면 'junior', 20 ~ 59이면 'senior', 60 이상이면 'veteran'으로 매핑.
