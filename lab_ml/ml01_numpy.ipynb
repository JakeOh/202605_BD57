## ravel vs flatten

*   ravel
    *   n차원 배열을 1차원 배열로 변환한 **뷰(view)** 를 반환.
    *   새로운 1차원 배열을 만드는 것이 아니라, 원본 배열의 모양을 1차원처럼 보여줄 뿐.
    *   반환된 1차원 배열에서 원소를 변경하면, 원본 배열의 원소도 변경됨!
*   flatten
    *   n차원 배열을 1차원 배열로 변환한 **복사본(copy)** 을 반환.
    *   원본 배열과는 별개의 새로운 1차원 배열을 메모리에 생성하고 반환.
    *   반환된 1차원 배열의 원소를 변경해도 원본 배열에는 아무 영향이 없음!

```python
arr_2d = np.arange(12).reshape((3, 4))
print(arr_2d)
```

```python
raveled = arr_2d.ravel()
print(raveled)
```

```python
flattened = arr_2d.flatten()
print(flattened)
```

```python
flattened[0] = 123  # 1차원 배열의 첫번째 원소를 123으로 변경
print(flattened)
```

```python
print(arr_2d)  #> flatten() 결과 배열을 수정하는 것은 원본 배열과는 아무런 상관이 없음.
```

```python
raveled[0] = 123
print(raveled)
```

```python
print(arr_2d)  #> ravel() 결과 배열을 수정하면 원본이 바뀜.
```

```python
arr_2d[0, 1] = 345
print(arr_2d)
```

```python
print(raveled)
```

# 배열 합치기(concatenate)

*   `numpy.concatenate()`, `numpy.concat()`: 특정 축(axis) 방향으로 배열을 이어붙이기.
*   `numpy.r_[array1, array2]`: 행(row)들을 이어붙이기. axis=0 방향으로 이어붙이기.
*   `numpy.c_[array1, array2]`: 열(column)들을 이어붙이기. axis=1 방향으로 이어붙이기.

```python
# (2, 3) shape의 2d array
arr1 = np.arange(6).reshape((2, 3))
print(arr1)
```

```python
# (1, 3) shape의 2d array
arr2 = np.array([[10, 20, 30]])
print(arr2)
```

```python
# (2, 1) shape의 2d array
arr3 = np.array([[100], [200]])
print(arr3)
```

```python
result = np.concat([arr1, arr2])  # axis 파라미터의 기본값은 0.
print(result)
```

```python
result = np.concat((arr1, arr3), axis=1)
print(result)
```

```python
np.r_[arr1, arr2]
```

```python
np.c_[arr1, arr3]
```

# Broadcasting

*   `numpy.ndarray`의 연산(`+, -, *, /, ==, !=, >, <, ...`)은 같은 인덱스(위치)의 원소들끼리(element-wise) 연산이 수행됨.
*   서로 모양(shape)이 다른 배열들 사이에서도 element-wise 연산이 가능한 경우가 있음.
*   broadcasting가 가능한 경우:
    *   누락되었거나 길이가 1인 차원에 대해서는 broadcasting이 가능.
    *   이어지는 차원의 축의 길이가 같은 경우에 broadcasting이 가능.

## numpy.ndarray와 scalar(값)의 연산

항상 연산이 가능(항상 broadcating이 가능).

스칼라를 배열의 각 원소마다 연산을 수행.

```python
arr = np.arange(1, 4)
print(arr)
print(arr + 10)
print(arr == 1)
```

```python
arr = np.arange(6).reshape((2, 3))
print(arr)
print(arr + 10)
```

## 2d array와 2d array의 broadcast

```python
a = np.arange(1, 7).reshape((2, 3))  # shape: (2, 3)
print(a)
```

```python
b = np.array([[7, 8, 9]])  # shape: (1, 3)
print(b)
```

```python
print(a + b)
```

```python
c = np.array([[7], [8]])  # shape: (2, 1)
print(c)
```

```python
print(a + c)
```

## 2d array와 1d array의 broadcast

```python
d = np.array([7, 8, 9])  # shape: (3,)
print(d)
```

```python
print(a + d)  # shape: (2, 3) + (3,)  --> broadcast가 가능한 경우.
```

```python
e = np.array([7, 8])  # shape: (2,)
print(e)
```

```python
print(a + e)  # shape: (2, 3) + (2,)  --> broadcast를 할 수 없는 경우.
```

## Broadcast 활용(예)

*   표준화(standardization): 평균이 0이 되고, 표준편차가 1이 되도록 값들을 변환.
>   $ z = \dfrac{x - \mu}{\sigma} $

*   정규화(normalization): 최솟값이 0이 되고, 최댓값이 1이 되도록 값들을 변환.
>   $ x_{norm} = \dfrac{x - x_{min}}{x_{max} - x_{min}} $

## 1차원 배열 표준화(standardization)

```python
np.random.seed(42)
x = np.random.randint(100, size=10)
print(x)
```

```python
mu = x.mean()  # 평균
print(mu)
```

```python
sigma = x.std()  # 표준편차(standard deviation)
print(sigma)
```

```python
x_scaled = (x - mu) / sigma
print(x_scaled)
```

```python
x_scaled.mean()  # 변환된 값들의 평균 = 0
```

```python
x_scaled.std()  # 변환된 값들의 표준편차 = 1
```

## 1차원 배열 정규화

```python
print(x)
```

```python
x_max = x.max()  # 최댓값
print(x_max)
```

```python
x_min = x.min()  # 최솟값
print(x_min)
```

```python
x_norm = (x - x_min) / (x_max - x_min)
print(x_norm)
```

## 2차원 배열에서의 표준화

```python
np.random.seed(42)
x = np.random.randint(100, size=(5, 3))  # shape: (5, 3)
print(x)
```

```python
x.mean()  # axis=None: 기본값. 모든 원소들의 평균 --> scalar
```

```python
x.mean(axis=0)  # axis=0 축 방향으로 평균을 계산 --> 3개의 평균값을 갖는 1차원 배열
```

```python
x.mean(axis=0, keepdims=True)  #> (1, 3) 모양의 2차원 배열.
```

```python
x.mean(axis=1)  # axis=1 축 방향으로 평균을 계산 --> 5개의 평균값을 갖는 1차원 배열
```

```python
x.mean(axis=1, keepdims=True)  #> (5, 1) 모양의 2차원 배열.
```

```python
x_scaled = (x - x.mean(axis=0, keepdims=True)) / x.std(axis=0, keepdims=True)
print(x_scaled)
```

```python
x_scaled.mean(axis=0)
```

```python
x_scaled.std(axis=0)
```

## 2차원 배열 정규화

```python
x_norm = (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0))
print(x_norm)
```
