## reshape

```python
arr_1d = np.arange(12)
print(arr_1d)
```

```python
# arr_1d 1차원 배열을 (4, 3) shape의 2차원 배열로 변환
arr_2d = arr_1d.reshape((4, 3))
print(arr_2d)
```

```python
# reshape 메서드에서 shape 아규먼트를 전달할 때, 자동으로 계산할 수 있는 차원의 값은 -1을 사용할 수 있음.
arr_2d = arr_1d.reshape((4, -1))
print(arr_2d)
```

```python
arr_2d = arr_1d.reshape((-1, 4))
print(arr_2d)
```

```python
arr_1d = np.arange(24)
print(arr_1d)
```

```python
# arr_1d를 (2, 3, 4) shape의 3차원 배열로 변환
arr_3d = arr_1d.reshape((2, 3, 4))
print(arr_3d)
```

```python
arr_3d = arr_1d.reshape((2, 3, -1))
print(arr_3d)
```

```python
arr_3d = arr_1d.reshape((-1, 3, 4))
print(arr_3d)
```

## ravel vs flatten

*   ravel
    *   n차원 배열을 1차원 배열로 변환한 **뷰(view)** 를 반환.
    *   새로운 1차원 배열을 만드는 것이 아니라, 원본 배열의 모양을 1차원처럼 보여줄 뿐.
    *   반환된 1차원 배열에서 원소를 변경하면, 원본 배열의 원소도 변경됨!
*   flatten
    *   n차원 배열을 1차원 배열로 변환한 **복사본(copy)** 을 반환.
    *   원본 배열과는 별개의 새로운 1차원 배열을 메모리에 생성하고 반환.
    *   반환된 1차원 배열의 원소를 변경해도 원본 배열에는 아무 영향이 없음!

```python
arr_2d = np.arange(12).reshape((3, 4))
print(arr_2d)
```

```python
raveled = arr_2d.ravel()
print(raveled)
```

```python
flattened = arr_2d.flatten()
print(flattened)
```

```python
flattened[0] = 123  # 1차원 배열의 첫번째 원소를 123으로 변경
print(flattened)
```

```python
print(arr_2d)  #> flatten() 결과 배열을 수정하는 것은 원본 배열과는 아무런 상관이 없음.
```

```python
raveled[0] = 123
print(raveled)
```

```python
print(arr_2d)  #> ravel() 결과 배열을 수정하면 원본이 바뀜.
```

```python
arr_2d[0, 1] = 345
print(arr_2d)
```

```python
print(raveled)
```

## `numpy.newaxis` 속성을 사용한 차원 늘리기

```python
arr = np.arange(6)
print(arr)
```

```python
arr.shape  #> (6,) shape - 1d arrary
```

```python
result = arr[np.newaxis, :]
print(result)
```

```python
result.shape  #> (1, 6) shape - 2d array
```

```python
arr.reshape((1, 6))
```

```python
arr.reshape((-1, 6))
```

```python
result = arr[:, np.newaxis]
print(result)
```

```python
result.shape  # (6, 1) shape - 2d array
```

```python
arr.reshape((6, 1))
```

```python
arr.reshape((6, -1))
```

# 배열 합치기(concatenate)

*   `numpy.concatenate()`, `numpy.concat()`: 특정 축(axis) 방향으로 배열을 이어붙이기.
*   `numpy.r_[array1, array2]`: 행(row)들을 이어붙이기. axis=0 방향으로 이어붙이기.
*   `numpy.c_[array1, array2]`: 열(column)들을 이어붙이기. axis=1 방향으로 이어붙이기.

```python
# (2, 3) shape의 2d array
arr1 = np.arange(6).reshape((2, 3))
print(arr1)
```

```python
# (1, 3) shape의 2d array
arr2 = np.array([[10, 20, 30]])
print(arr2)
```

```python
# (2, 1) shape의 2d array
arr3 = np.array([[100], [200]])
print(arr3)
```

```python
result = np.concat([arr1, arr2])  # axis 파라미터의 기본값은 0.
print(result)
```

```python
result = np.concat((arr1, arr3), axis=1)
print(result)
```

```python
np.r_[arr1, arr2]
```

```python
np.c_[arr1, arr3]
```

# Broadcasting

*   `numpy.ndarray`의 연산(`+, -, *, /, ==, !=, >, <, ...`)은 같은 인덱스(위치)의 원소들끼리(element-wise) 연산이 수행됨.
*   서로 모양(shape)이 다른 배열들 사이에서도 element-wise 연산이 가능한 경우가 있음.
*   broadcasting가 가능한 경우:
    *   누락되었거나 길이가 1인 차원에 대해서는 broadcasting이 가능.
    *   이어지는 차원의 축의 길이가 같은 경우에 broadcasting이 가능.

## numpy.ndarray와 scalar(값)의 연산

항상 연산이 가능(항상 broadcating이 가능).

스칼라를 배열의 각 원소마다 연산을 수행.

```python
arr = np.arange(1, 4)
print(arr)
print(arr + 10)
print(arr == 1)
```

```python
arr = np.arange(6).reshape((2, 3))
print(arr)
print(arr + 10)
```
