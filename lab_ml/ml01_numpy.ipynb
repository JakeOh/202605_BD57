# `numpy.ndarray` 클래스의 속성들(properties)

```python
# 1차원 배열
arr = np.arange(10)
print(arr)
arr
```

```python
print('shape:', arr.shape)  #> tuple
print('size:', arr.size)  #> 배열의 모든 아이템의 개수.
print('ndim:', arr.ndim)  #> dimension(차원) - 축(axis)의 개수.
print('dtype:', arr.dtype)
```

```python
# 2차원 배열
arr = np.ones(shape=(2, 3))
print(arr)
```

```python
print('shape:', arr.shape)  #> tuple
print('size:', arr.size)  #> 배열의 모든 아이템의 개수.
print('ndim:', arr.ndim)  #> dimension(차원) - 축(axis)의 개수.
print('dtype:', arr.dtype)
```

```python
# 3차원 배열(2차원 배열들을 원소로 갖는 배열)
arr = np.arange(1, 25).reshape((2, 3, 4))
print(arr)
```

```python
print(arr[0])
```

```python
print(arr[0, 0])  # print(arr[0][0])
```

```python
print(arr[0, 0, 0])  # print(arr[0][0][0])
```

```python
print('shape:', arr.shape)  #> tuple
print('size:', arr.size)  #> 배열의 모든 아이템의 개수.
print('ndim:', arr.ndim)  #> dimension(차원) - 축(axis)의 개수.
print('dtype:', arr.dtype)
```

# Slicing

## 1차원 배열의 slicing

```python
arr = np.arange(10)
print(arr)
```

```python
arr[1:5]
```

```python
arr[:3]  # 시작 인덱스 0은 생략 가능
```

```python
arr[-3:]  # 배열의 끝까지 자를 때는 마지막 인덱스를 생략할 수 있음.
```

```python
print(arr[0:1])  #> slicing --> 배열
print(arr[0])    #> indexing --> 배열의 원소
```

## 2차원 배열의 slicing

```python
arr = np.arange(20).reshape((4, 5))
print(arr)
```

```python
print(arr[:2])
```

```python
print(arr[:2, :])
```

```python
print(arr[:2, :3])
```

```python
print(arr[1:3, 1:4])
```

## 3차원 배열의 slicing

```python
arr = np.arange(60).reshape((3, 4, 5))
print(arr)
```

```python
print(arr[:2])
```

```python
print(arr[:2, :, :])
```

```python
print(arr[:2, :2])
```

```python
print(arr[:2, :2, :])
```

```python
print(arr[:2, :2, :2])
```

```python
print(arr[:1, :2, :2])  #> 3차원 배열
```

```python
print(arr[0, :2, :2])  #> 2차원 배열
```
