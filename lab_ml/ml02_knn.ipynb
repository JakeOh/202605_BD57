# Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# KNN(k-Nearest Neighbor) 알고리즘 분류기(classifier) 클래스
from sklearn.neighbors import KNeighborsClassifier
```

# 데이터 준비

```python
file_path = 'https://bit.ly/fish_csv_data'  # github의 csv 파일 주소
```

```python
fish = pd.read_csv(file_path)
```

```python
fish.head()
```

```python
fish.info()
```

# 데이터 탐색

## 카테고리 변수 탐색

```python
fish.Species.value_counts()  # 물고기 종류 빈도수
```

```python
sns.countplot(data=fish, y='Species',
              order=fish.Species.value_counts().index)
plt.show()
```

## 연속형 변수들 탐색

```python
fish.describe()
```

연속형 변수들의 histogram과 box plot을 시각화

```python
features = fish.columns[1:]
print(features)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=5, figsize=(20, 8))

for i, var in enumerate(features):
    sns.boxplot(data=fish, x=var, ax=axes[0, i])
    sns.histplot(data=fish, x=var, ax=axes[1, i])

plt.show()
```
